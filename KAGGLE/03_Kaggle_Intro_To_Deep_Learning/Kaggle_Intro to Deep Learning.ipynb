### What is Deep Learning?

Some of the most impressive advances in artificial intelligence in recent years have been in the field of deep learning. [ Son yıllarda yapay zekadaki en etkileyici ilerlemelerden bazıları derin öğrenme alanında olmuştur.] Natural language translation, image recognition, and game playing are all tasks where deep learning models have neared or even exceeded human-level performance. [ Doğal dil çevirisi, görüntü tanıma ve oyun oynama, derin öğrenme modellerinin insan düzeyinde performansa yaklaştığı veya hatta aştığı görevlerdir.]

So what is deep learning? [ Peki derin öğrenme nedir?] Deep learning is an approach to machine learning characterized by deep stacks of computations. [ Derin öğrenme, derin hesaplama yığınları ile karakterize edilen makine öğrenimine bir yaklaşımdır.] This depth of computation is what has enabled deep learning models to disentangle the kinds of complex and hierarchical patterns found in the most challenging real-world datasets. [ Bu hesaplama derinliği, derin öğrenme modellerinin en zorlu gerçek dünya veri kümelerinde bulunan karmaşık ve hiyerarşik kalıpları çözmesini sağlayan şeydir.]

Through their power and scalability neural networks have become the defining model of deep learning. [ Güçleri ve ölçeklenebilirliği sayesinde sinir ağları, derin öğrenmenin tanımlayıcı modeli haline gelmiştir.] Neural networks are composed of neurons, where each neuron individually performs only a simple computation. [ Nöral ağlar nöronlardan oluşur, burada her nöron tek tek basit bir hesaplama gerçekleştirir.] The power of a neural network comes instead from the complexity of the connections these neurons can form. [ Bir sinir ağının gücü, bu nöronların oluşturabileceği bağlantıların karmaşıklığından gelir.]

### The Linear Unit

So let's beg [ Bir sinir ağı ağırlıklarını değiştirerek 'öğrenir'.] in with the fundamental component of a neural network: the individual neuron. [ Öyleyse bir sinir ağının temel bileşeni ile başlayalım: bireysel nöron.] As a diagram, a neuron (or unit) with one input looks like: [ Bir [ içinde] diyagram olarak, bir girişli bir nöron (veya birim):]



The input is x. [ Giriş x.] Its connection to the neuron has a weight which is w. [ Nöron bağlantısı w olan bir ağırlığa sahiptir.] Whenever a value flows through a connection, you multiply the value by the connection's weight. [ Bir değer bir bağlantıdan ne zaman akarsa, değeri bağlantının ağırlığına göre çarparsınız.] For the input x, what reaches the neuron is w * x. [ X girişi için nörona ulaşan şey w * x'dir.] A neural network "learns" by modifying its weights. [ Bir sinir ağı ağırlıklarını değiştirerek 'öğrenir'.]

The b is a special kind of weight we call the bias. [ B, önyargı dediğimiz özel bir ağırlık türüdür.] The bias doesn't have any input data associated with it; instead, we put a 1 in the diagram so that the value that reaches the neuron is just b (since 1 * b = b). [ Önyargı, onunla ilişkili herhangi bir giriş verisi yoktur; Bunun yerine, nörona ulaşan değer sadece B (1 * b = b) olacak şekilde diyagrama 1 koyduk.] The bias enables the neuron to modify the output independently of its inputs. [ Önyargı, nöronun çıktıyı girişlerinden bağımsız olarak değiştirmesini sağlar.]

The y is the value the neuron ultimately outputs. [ Y, nöronun sonuçta çıkttığı değerdir.] To get the output, the neuron sums up all the values it receives through its connections. [ Çıktıyı elde etmek için nöron, bağlantıları aracılığıyla aldığı tüm değerleri özetler.] This neuron's activation is y = w * x + b, or as a formula  y=wx+b . [ Bu nöronun aktivasyonu y = w * x+b veya bir formül olarak y = wx+by = wx+b.]



### Example - The Linear Unit as a Model

Though individual neurons will usually only function as part of a larger network, it's often useful to start with a single neuron model as a baseline. [ Bireysel nöronlar genellikle sadece daha büyük bir ağın parçası olarak işlev görse de, temel olarak tek bir nöron modeliyle başlamak genellikle yararlıdır.] Single neuron models are linear models. [ Tek nöron modelleri doğrusal modellerdir.]

Let's think about how this might work on a dataset like 80 Cereals. [ Bunun 80 tahıl gibi bir veri kümesinde nasıl çalışabileceğini düşünelim.] Training a model with 'sugars' (grams of sugars per serving) as input and 'calories' (calories per serving) as output, we might find the bias is b=90 and the weight is w=2.5. [ Giriş ve 'kalori' (porsiyon başına kalori) olarak 'şeker' (porsiyon başına şeker gramı) ile bir model eğitimi çıktı olarak, önyargının B = 90 ve ağırlığın W = 2.5 olduğunu görebiliriz.] We could estimate the calorie content of a cereal with 5 grams of sugar per serving like this: [ Hizmet başına 5 gram şeker ile bir tahılın kalori içeriğini tahmin edebiliriz:]



And, checking against our formula, we have  calories=2.5×5+90=102.5 , just like we expect. [ Ve formülümüze karşı kontrol ederek, tıpkı beklediğimiz gibi kalorilerimiz = 2.5 × 5+90 = 102.5 kalori = 2.5 × 5+90 = 102.5 var]

### Multiple Inputs

The 80 Cereals dataset has many more features than just 'sugars'. [ 80 Cereals veri kümesinin sadece 'şekerlerden' çok daha fazla özelliği vardır.] What if we wanted to expand our model to include things like fiber or protein content? [ Modelimizi lif veya protein içeriği gibi şeyleri içerecek şekilde genişletmek istersek ne olur?] That's easy enough. [ Bu yeterince kolay.] We can just add more input connections to the neuron, one for each additional feature. [ Her ek özellik için bir tane olmak üzere nörona daha fazla giriş bağlantısı ekleyebiliriz.] To find the output, we would multiply each input to its connection weight and then add them all together. [ Çıktıyı bulmak için, her girişi bağlantı ağırlığına çarpar ve sonra hepsini bir araya getiririz.]



The formula for this neuron would be  y=w0x0+w1x1+w2x2+b . [ Bu nöronun formülü y = w0x0+w1x1+w2x2+by = w0x0+w1x1+w2x2+b olacaktır.] A linear unit with two inputs will fit a plane, and a unit with more inputs than that will fit a hyperplane. [ İki girişe sahip doğrusal bir ünite bir düzleme uyacaktır ve bundan daha fazla girişe sahip bir ünite bir hiper tablaya uyacaktır.]

### Linear Units in Keras

The easiest way to create a model in Keras is through keras.Sequential, which creates a neural network as a stack of layers. [ Keras'ta bir model oluşturmanın en kolay yolu Keras'dır.] We can create models like those above using a dense layer (which we'll learn more about in the next lesson). [ Yukarıdaki gibi modeller oluşturabiliriz, yoğun bir katman kullanarak (bir sonraki derste daha fazla bilgi edineceğiz).]

We could define a linear model accepting three input features ('sugars', 'fiber', and 'protein') and producing a single output ('calories') like so: [ Üç giriş özelliğini ('şeker', 'fiber' ve 'protein') kabul eden ve bunun gibi tek bir çıkış ('kalori') üreten doğrusal bir model tanımlayabiliriz:]

```python
from tensorflow import keras
from tensorflow.keras import layers

# Create a network with 1 linear unit
model = keras.Sequential([
                            layers.Dense(units=1, input_shape=[3])
                        ])
```

With the first argument, units, we define how many outputs we want. [ İlk argüman, birimlerle, kaç çıktı istediğimizi tanımlarız.] In this case we are just predicting 'calories', so we'll use units=1. [ Bu durumda sadece 'kalorileri' tahmin ediyoruz, bu yüzden birimler = 1 kullanacağız.]

With the second argument, input_shape, we tell Keras the dimensions of the inputs. [ İkinci argüman olan Input_Shape ile Keras'a girişlerin boyutlarını söylüyoruz.] Setting input_shape=[3] ensures the model will accept three features as input ('sugars', 'fiber', and 'protein'). [ Input_shape = [3] ayarlama, modelin üç özelliği giriş ('şeker', 'fiber' ve 'protein') olarak kabul etmesini sağlar.]

This model is now ready to be fit to training data! [ Bu model artık eğitim verilerine uygun olmaya hazır!]

## 2. Deep Neural Networks

### Introduction

In this lesson we're going to see how we can build neural networks capable of learning the complex kinds of relationships deep neural nets are famous for. [ Bu derste, derin sinir ağlarının ünlü olduğu karmaşık ilişkileri öğrenebilen sinir ağları nasıl oluşturabileceğimizi göreceğiz.]

The key idea here is modularity, building up a complex network from simpler functional units. [ Buradaki temel fikir, daha basit fonksiyonel birimlerden karmaşık bir ağ oluşturan modülerliktir.] We've seen how a linear unit computes a linear function -- now we'll see how to combine and modify these single units to model more complex relationships. [ Doğrusal bir ünitenin doğrusal bir işlevi nasıl hesapladığını gördük - şimdi daha karmaşık ilişkileri modellemek için bu tek birimleri nasıl birleştireceğimizi ve değiştireceğimizi göreceğiz.]

### Layers

Neural networks typically organize their neurons into layers. [ Sinir ağları tipik olarak nöronlarını katmanlara dönüştürür.] When we collect together linear units having a common set of inputs we get a dense layer. [ Ortak bir giriş setine sahip doğrusal birimleri bir araya getirdiğimizde yoğun bir katman elde ederiz.]



You could think of each layer in a neural network as performing some kind of relatively simple transformation. [ Bir sinir ağındaki her katmanı bir tür nispeten basit dönüşüm gerçekleştirerek düşünebilirsiniz.] Through a deep stack of layers, a neural network can transform its inputs in more and more complex ways. [ Derin bir katman yığını sayesinde, bir sinir ağı girdilerini daha karmaşık şekillerde dönüştürebilir.] In a well-trained neural network, each layer is a transformation getting us a little bit closer to a solution. [ İyi eğitimli bir sinir ağında, her katman bizi bir çözüme biraz daha yaklaştıran bir dönüşümdür.]

### The Activation Function¶

It turns out, however, that two dense layers with nothing in between are no better than a single dense layer by itself. [ Bununla birlikte, aralarında hiçbir şey olmayan iki yoğun katmanın tek bir yoğun katmandan daha iyi olmadığı ortaya çıkıyor.] Dense layers by themselves can never move us out of the world of lines and planes. [ Yoğun katmanlar bizi asla çizgiler ve uçaklar dünyasından uzaklaştıramaz.] What we need is something nonlinear. [ İhtiyacımız olan şey doğrusal olmayan bir şey.] What we need are activation functions. [ İhtiyacımız olan aktivasyon işlevleri.]



An **activation function** is simply some function we apply to each of a layer's outputs (its activations). [ Bir aktivasyon işlevi, bir katmanın çıkışlarının her birine (aktivasyonları) uyguladığımız bir işlevdir.] The most common is the rectifier function  max(0,x) . [ En yaygın olanı, maks (0, x) maks (0, x) doğrultucu fonksiyonudur.]



The rectifier function has a graph that's a line with the negative part "rectified" to zero. [ Doğrultucu işlevi, negatif parça 'düzeltilmiş' olan bir çizgi olan bir grafiğe sahiptir.] Applying the function to the outputs of a neuron will put a bend in the data, moving us away from simple lines. [ İşlevi bir nöronun çıkışlarına uygulamak, verilere bir viraj koyarak bizi basit çizgilerden uzaklaştıracaktır.]

When we attach the rectifier to a linear unit, we get a rectified linear unit or ReLU. [ Doğrultayı doğrusal bir üniteye taktığımızda, düzeltilmiş bir doğrusal ünite veya relu alırız.] (For this reason, it's common to call the rectifier function the "ReLU function".) [ (Bu nedenle, doğrultucu işlevini 'relu işlevi' olarak adlandırmak yaygındır.)] Applying a ReLU activation to a linear unit means the output becomes max(0, w * x + b), which we might draw in a diagram like: [ Doğrusal bir üniteye bir relu aktivasyonu uygulamak, çıkışın maksimum (0, W * x + b) olduğu anlamına gelir;]



### Stacking Dense Layers

Now that we have some nonlinearity, let's see how we can stack layers to get complex data transformations. [ Artık bazı doğrusal olmayanlığa sahip olduğumuza göre, karmaşık veri dönüşümleri elde etmek için katmanları nasıl istifleyebileceğimizi görelim.]



The layers before the output layer are sometimes called hidden since we never see their outputs directly. [ Çıkış katmanından önceki katmanlara bazen gizli olarak adlandırılır, çünkü çıkışlarını asla doğrudan görmeziz.]
Now, notice that the final (output) layer is a linear unit (meaning, no activation function). [ Şimdi, nihai (çıkış) katmanının doğrusal bir birim olduğuna dikkat edin (yani aktivasyon işlevi yok).] That makes this network appropriate to a regression task, where we are trying to predict some arbitrary numeric value. [ Bu, bu ağı, keyfi sayısal değeri tahmin etmeye çalıştığımız bir regresyon görevine uygun hale getirir.] Other tasks (like classification) might require an activation function on the output. [ Diğer görevler (sınıflandırma gibi) çıktı üzerinde bir aktivasyon işlevi gerektirebilir.]

#### Building Sequential Models [ Sıralı modeller oluşturma] 

The Sequential model we've been using will connect together a list of layers in order from first to last: the first layer gets the input, the last layer produces the output. [ Kullandığımız sıralı model, birinci kattan sonuna kadar bir katman listesi birbirine bağlayacaktır: ilk katman girişi alır, son katman çıkışı üretir.] This creates the model in the figure above: [ Bu, yukarıdaki şekilde modeli oluşturur:]

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
                            # the hidden ReLU layers
                            layers.Dense(units=4, activation='relu', input_shape=[2]),
                            layers.Dense(units=3, activation='relu'),
                            # the linear output layer 
                            layers.Dense(units=1),
                        ])

# Be sure to pass all the layers together in a list, like [layer, layer, layer, ...], instead of as separate arguments. 
# To add an activation function to a layer, just give its name in the activation argument.
```

## 3. Stochastic Gradient Descent

### Introduction

In the first two lessons, we learned how to build fully-connected networks out of stacks of dense layers. [ İlk iki dersde, yoğun katman yığınlarından tamamen bağlantılı ağların nasıl oluşturulacağını öğrendik.] When first created, all of the network's weights are set randomly -- the network doesn't "know" anything yet. [ İlk oluşturulduğunda, ağın tüm ağırlıkları rastgele ayarlanır - ağ henüz hiçbir şey bilmiyor.] In this lesson we're going to see how to train a neural network; we're going to see how neural networks learn. [ Bu derste bir sinir ağının nasıl eğitileceğini göreceğiz; Sinir ağlarının nasıl öğrendiğini göreceğiz.]

As with all machine learning tasks, we begin with a set of training data. [ Tüm makine öğrenimi görevlerinde olduğu gibi, bir dizi eğitim verisi ile başlıyoruz.] Each example in the training data consists of some features (the inputs) together with an expected target (the output). [ Eğitim verilerindeki her örnek, beklenen bir hedefle (çıktı) birlikte bazı özelliklerden (girişler) oluşur.] Training the network means adjusting its weights in such a way that it can transform the features into the target. [ Ağın eğitimi, ağırlıklarını özellikleri hedefe dönüştürebilecek şekilde ayarlamak anlamına gelir.] In the 80 Cereals dataset, for instance, we want a network that can take each cereal's 'sugar', 'fiber', and 'protein' content and produce a prediction for that cereal's 'calories'. [ Örneğin 80 Cereals veri kümesinde, her tahılın 'şeker', 'lif' ve 'protein' içeriğini alabilen ve o tahılın 'kalorileri' için bir tahmin üretebilecek bir ağ istiyoruz.] If we can successfully train a network to do that, its weights must represent in some way the relationship between those features and that target as expressed in the training data. [ Bunu yapmak için bir ağı başarıyla eğitebilirsek, ağırlıkları bir şekilde bu özellikler ile bu hedef arasındaki ilişkiyi eğitim verilerinde ifade edildiği gibi temsil etmelidir.]

In addition to the training data, we need two more things: [ Eğitim verilerine ek olarak, iki şeye daha ihtiyacımız var:]

- A "loss function" that measures how good the network's predictions are. [ Ağın tahminlerinin ne kadar iyi olduğunu ölçen bir 'kayıp işlevi'.]

- An "optimizer" that can tell the network how to change its weights. [ Ağa ağırlıklarını nasıl değiştireceğini söyleyebilen bir 'optimizer'.]

### The Loss Function

We've seen how to design an architecture for a network, but we haven't seen how to tell a network what problem to solve. [ Bir ağ için bir mimarinin nasıl tasarlanacağını gördük, ancak bir ağın nasıl çözüleceğini nasıl çözeceğimizi görmedik.] This is the job of the loss function. [ Bu kayıp fonksiyonunun işidir.]

The **loss function** measures the disparity between the the target's true value and the value the model predicts. [ Kayıp fonksiyonu, hedefin gerçek değeri ile modelin öngördüğü değer arasındaki eşitsizliği ölçer.]

Different problems call for different loss functions. [ Farklı sorunlar farklı kayıp fonksiyonları gerektirir.] We have been looking at **regression** problems, where the task is to predict some numerical value -- calories in 80 Cereals, rating in Red Wine Quality. [ Görevin bazı sayısal değeri tahmin etmek olduğu regresyon problemlerine bakıyoruz - 80 tahıllarda kalori, kırmızı şarap kalitesinde derecelendirme.] Other regression tasks might be predicting the price of a house or the fuel efficiency of a car. [ Diğer regresyon görevleri bir evin fiyatını veya bir arabanın yakıt verimliliğini tahmin ediyor olabilir.]

A common loss function for regression problems is the **mean absolute error** or **MAE**. [ Regresyon problemleri için yaygın bir kayıp fonksiyonu ortalama mutlak hata veya MAE'dir.] For each prediction y_pred, MAE measures the disparity from the true target y_true by an absolute difference abs(y_true - y_pred). [ Her tahmini y_pred için Mae, mutlak bir fark ABS (y_true - y_pred) ile gerçek hedef y_true'den eşitsizliği ölçer.]

The total MAE loss on a dataset is the mean of all these absolute differences. [ Bir veri kümesindeki toplam MAE kaybı, tüm bu mutlak farklılıkların ortalamasıdır.]



Besides MAE, other loss functions you might see for regression problems are the mean-squared error (MSE) or the Huber loss (both available in Keras). [ MAE'nin yanı sıra, regresyon problemleri için görebileceğiniz diğer kayıp fonksiyonları ortalama kare hatası (MSE) veya Huber kaybıdır (her ikisi de Keras'ta mevcuttur).]

During training, the model will use the loss function as a guide for finding the correct values of its weights (lower loss is better). [ Eğitim sırasında, model kayıp işlevini ağırlıklarının doğru değerlerini bulmak için bir kılavuz olarak kullanacaktır (daha düşük kayıp daha iyidir).] In other words, the loss function tells the network its objective. [ Başka bir deyişle, kayıp fonksiyonu ağa hedefini söyler.]

### The Optimizer - Stochastic Gradient Descent 

We've described the problem we want the network to solve, but now we need to say how to solve it. [ Ağın çözmesini istediğimiz sorunu açıkladık, ancak şimdi nasıl çözüleceğini söylemeliyiz.] This is the job of the **optimizer**. [ Bu optimize edicinin işi.] The optimizer is an algorithm that adjusts the weights to minimize the loss. [ Optimize edici, kaybı en aza indirmek için ağırlıkları ayarlayan bir algoritmadır.]

Virtually all of the optimization algorithms used in deep learning belong to a family called **stochastic gradient descent**. [ Derin öğrenmede kullanılan neredeyse tüm optimizasyon algoritmaları stokastik gradyan iniş adı verilen bir aileye aittir.] They are iterative algorithms that train a network in steps. [ Bir ağı adımlarla eğiten yinelemeli algoritmalardır.] One **step** of training goes like this: [ Bir antrenman adımı şöyle gidiyor:]

1. Sample some training data and run it through the network to make predictions. [ Bazı eğitim verilerini örnekleyin ve tahmin yapmak için ağdan geçirin.]

2. Measure the loss between the predictions and the true values. [ Tahminler ve gerçek değerler arasındaki kaybı ölçün.]

3. Finally, adjust the weights in a direction that makes the loss smaller. [ Son olarak, ağırlıkları kaybı küçülten bir yönde ayarlayın.]

Then just do this over and over until the loss is as small as you like (or until it won't decrease any further.) [ O zaman kayıp istediğiniz kadar küçük olana kadar bunu tekrar tekrar yapın (veya daha fazla düşmeyene kadar).]



Each iteration's sample of training data is called a **minibatch** (or often just "batch"), while a complete round of the training data is called an **epoch**. [ Her yinelemenin eğitim verileri örneğine minibatch (veya genellikle sadece 'toplu') denirken, eğitim verilerinin tam bir turuna çağ denir.] The number of epochs you train for is how many times the network will see each training example. [ Eğittiğiniz çağların sayısı, ağın her eğitim örneğini kaç kez göreceğidir.]

The animation shows the linear model from Lesson 1 being trained with SGD. [ Animasyon, SGD ile eğitilen Ders 1'deki doğrusal modeli göstermektedir.] The pale red dots depict the entire training set, while the solid red dots are the minibatches. [ Soluk kırmızı noktalar tüm eğitim setini tasvir ederken, katı kırmızı noktalar minibatches.] Every time SGD sees a new minibatch, it will shift the weights (w the slope and b the y-intercept) toward their correct values on that batch. [ SGD yeni bir minibatch her gördüğünde, ağırlıkları (eğim ve y kesişimi w) o parti üzerindeki doğru değerlerine kaydırır.] Batch after batch, the line eventually converges to its best fit. [ Parti partiden sonra, çizgi sonunda en iyi uyumuna yaklaşır.] You can see that the loss gets smaller as the weights get closer to their true values. [ Ağırlıklar gerçek değerlerine yaklaştıkça kaybın azaldığını görebilirsiniz.]

### Learning Rate and Batch Size 

Notice that the line only makes a small shift in the direction of each batch (instead of moving all the way). [ Çizginin yalnızca her bir parti yönünde küçük bir kayma yaptığına dikkat edin (sonuna kadar hareket etmek yerine).] The size of these shifts is determined by the **learning rate**. [ Bu vardiyaların boyutu öğrenme oranı ile belirlenir.] A smaller learning rate means the network needs to see more minibatches before its weights converge to their best values. [ Daha küçük bir öğrenme oranı, ağın ağırlıkları en iyi değerlerine yaklaşmadan önce daha fazla minibatch görmesi gerektiği anlamına gelir.]

The learning rate and the size of the minibatches are the two parameters that have the largest effect on how the SGD training proceeds. [ Öğrenme oranı ve minibatches boyutu, SGD eğitiminin nasıl ilerlediği üzerinde en büyük etkiye sahip iki parametredir.] Their interaction is often subtle and the right choice for these parameters isn't always obvious. [ Etkileşimleri genellikle incedir ve bu parametreler için doğru seçim her zaman açık değildir.] (We'll explore these effects in the exercise.) [ (Egzersizde bu etkileri keşfedeceğiz.)]

Fortunately, for most work it won't be necessary to do an extensive hyperparameter search to get satisfactory results. [ Neyse ki, çoğu iş için tatmin edici sonuçlar elde etmek için kapsamlı bir hiperparametre arama yapmak gerekli olmayacaktır.] **Adam** is an SGD algorithm that has an adaptive learning rate that makes it suitable for most problems without any parameter tuning (it is "self tuning", in a sense). [ Adam, herhangi bir parametre ayarlaması olmadan çoğu problem için uygun hale getiren uyarlanabilir bir öğrenme oranına sahip bir SGD algoritmasıdır (bir anlamda 'kendi kendini ayarlama').] Adam is a great general-purpose optimizer. [ Adam büyük bir genel amaçlı optimize edicidir.]

### Adding the Loss and Optimizer

After defining a model, you can add a loss function and optimizer with the model's compile method: [ Bir modeli tanımladıktan sonra, modelin derleme yöntemiyle bir kayıp işlevi ve optimize edici ekleyebilirsiniz:]

model.compile(  
                optimizer="adam", 
                loss="mae", 
            )
Notice that we are able to specify the loss and optimizer with just a string. [ Kaybı ve optimize ediciyi yalnızca bir dize ile belirtebileceğimize dikkat edin.] You can also access these directly through the Keras API -- if you wanted to tune parameters, for instance -- but for us, the defaults will work fine.

**What's In a Name?**
The **gradient** is a vector that tells us in what direction the weights need to go. [ Gradyan, ağırlıkların hangi yönde gitmesi gerektiğini söyleyen bir vektördür.] More precisely, it tells us how to change the weights to make the loss change fastest. [ Daha doğrusu, kaybın en hızlı değişmesi için ağırlıkların nasıl değiştirileceğini anlatıyor.] We call our process gradient **descent** because it uses the gradient to descend the loss curve towards a minimum. [ İşlem gradyanımımızı çağırıyoruz, çünkü kayıp eğrisini minimumda inmek için gradyanı kullanıyor.] **Stochastic** means "determined by chance." [ Stokastik 'şans eseri belirlenir' anlamına gelir.] Our training is stochastic because the minibatches are random samples from the dataset. [ Eğitimimiz stokastiktir çünkü minibatches veri kümesinden rastgele örneklerdir.] And that's why it's called SGD! [ Ve bu yüzden buna SGD deniyor!]

### Example - Red Wine Quality

Now we know everything we need to start training deep learning models. [ Artık derin öğrenme modellerini eğitmeye başlamak için ihtiyacımız olan her şeyi biliyoruz.] So let's see it in action! [ Öyleyse bakalım eylemde!] We'll use the Red Wine Quality dataset. [ Kırmızı şarap kalitesi veri kümesini kullanacağız.]

This dataset consists of physiochemical measurements from about 1600 Portuguese red wines. [ Bu veri kümesi, yaklaşık 1600 Portekizli kırmızı şaraptan fizyokimyasal ölçümlerden oluşmaktadır.] Also included is a quality rating for each wine from blind taste-tests. [ Ayrıca kör tat testlerinden her şarap için kaliteli bir derecelendirme de dahildir.] How well can we predict a wine's perceived quality from these measurements? [ Bu ölçümlerden bir şarabın algılanan kalitesini ne kadar iyi tahmin edebiliriz?]

We've put all of the data preparation into this next hidden cell. [ Tüm veri hazırlamalarını bir sonraki gizli hücreye koyduk.] It's not essential to what follows so feel free to skip it. [ Aşağıdaki şey için gerekli değildir, bu yüzden atlamaktan çekinmeyin.] One thing you might note for now though is that we've rescaled each feature to lie in the interval  [0,1] . [ Şimdilik not edebileceğiniz bir şey, her özelliği [0,1] [0,1] aralığında yatmak için yeniden ölçeklendirmiş olmamızdır.] As we'll discuss more in Lesson 5, neural networks tend to perform best when their inputs are on a common scale. [ Ders 5'te daha fazla tartışacağımız gibi, sinir ağları, girdileri ortak ölçekte olduğunda en iyi performans gösterme eğilimindedir.]

```python
import pandas as pd
import numpy as np
from IPython.display import display

red_wine = pd.read_csv('winequality-red.csv')

# Create training and validation splits
df_train = red_wine.sample(frac=0.7, random_state=0)
df_valid = red_wine.drop(df_train.index)
display(df_train.head(4))

# Scale to [0, 1]
max_ = df_train.max(axis=0)
min_ = df_train.min(axis=0)
df_train = (df_train - min_) / (max_ - min_)
df_valid = (df_valid - min_) / (max_ - min_)

# Split features and target
X_train = df_train.drop('quality', axis=1)
X_valid = df_valid.drop('quality', axis=1)
y_train = df_train['quality']
y_valid = df_valid['quality']
```

How many inputs should this network have? [ Bu ağın kaç girişi olmalı?] We can discover this by looking at the number of columns in the data matrix. [ Veri matrisindeki sütun sayısına bakarak bunu keşfedebiliriz.] Be sure not to include the target ('quality') here -- only the input features. [ Hedefi ('kalite') buraya eklemediğinizden emin olun - yalnızca giriş özellikleri.]

```python
print(X_train.shape)
```

Eleven columns means eleven inputs. [ On bir sütun on bir giriş anlamına gelir.]
We've chosen a three-layer network with over 1500 neurons. [ 1500'den fazla nörona sahip üç katmanlı bir ağ seçtik.] This network should be capable of learning fairly complex relationships in the data. [ Bu ağ verilerde oldukça karmaşık ilişkiler öğrenebilmelidir.]

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
                            layers.Dense(512, activation='relu', input_shape=[11]),
                            layers.Dense(512, activation='relu'),
                            layers.Dense(512, activation='relu'),
                            layers.Dense(1),
                        ])
```

Deciding the architecture of your model should be part of a process. [ Modelinizin mimarisine karar vermek bir sürecin parçası olmalıdır.] Start simple and use the validation loss as your guide. [ Basit başlayın ve doğrulama kaybını rehberiniz olarak kullanın.] You'll learn more about model development in the exercises. [ Egzersizlerde model gelişimi hakkında daha fazla bilgi edineceksiniz.]

After defining the model, we compile in the optimizer and loss function. [ Modeli tanımladıktan sonra, optimize edici ve kayıp işlevini derleriz.]

```python
model.compile(
              optimizer='adam',
              loss='mae',
             )
```

Now we're ready to start the training! [ Şimdi eğitime başlamaya hazırız!] We've told Keras to feed the optimizer 256 rows of the training data at a time (the batch_size) and to do that 10 times all the way through the dataset (the epochs). [ Keras'a bir seferde (Batch_size) optimize edici 256 sırasını beslemesini ve bunu veri kümesinden (dönemler) 10 kez yapmasını söyledik.]

```python
history = model.fit(
                    X_train, y_train,
                    validation_data=(X_valid, y_valid),
                    batch_size=256,
                    epochs=10,
                    )
```

You can see that Keras will keep you updated on the loss as the model trains. [ Model trenler olarak Keras'ın kayıp konusunda sizi güncel tutacağını görebilirsiniz.]

Often, a better way to view the loss though is to plot it. [ Genellikle, kaybı görüntülemenin daha iyi bir yolu onu planlamaktır.] The fit method in fact keeps a record of the loss produced during training in a History object. [ Uyum yöntemi aslında bir tarih nesnesinde eğitim sırasında üretilen kaybın kaydını tutar.] We'll convert the data to a Pandas dataframe, which makes the plotting easy. [ Verileri bir Pandas veri çerçevesine dönüştüreceğiz, bu da çizimi kolaylaştırır.]

```python
import pandas as pd

# convert the training history to a dataframe
history_df = pd.DataFrame(history.history)
# use Pandas native plot method
history_df['loss'].plot();
```

Notice how the loss levels off as the epochs go by. [ Çağlar geçtikçe kayıp seviyelerinin nasıl düştüğüne dikkat edin.] When the loss curve becomes horizontal like that, it means the model has learned all it can and there would be no reason continue for additional epochs. [ Kayıp eğrisi bu şekilde yatay hale geldiğinde, modelin elinden geleni öğrendiği ve ek dönemler için devam etmeyeceği anlamına gelir.]
