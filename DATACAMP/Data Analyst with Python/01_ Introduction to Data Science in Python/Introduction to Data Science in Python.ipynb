It's easy to get errors when working with strings. [Dizelerle çalışırken hata almak kolaydır.] If you get one, there are two likely causes. [Bunu yaparsanız Python, dizginizin bir değişken olduğunu düşünecek ve bu değişken daha önce tanımlanmamışsa bir 'ad hatası' alacaksınız.] Alternatively, you might have started your string with one type of quote and finished with a different type. [Alternatif olarak, dizenize bir tür alıntıyla başlayıp farklı bir türle bitirmiş olabilirsiniz.] If you mix single and double quotes, you'll get a syntax error. [Tek ve çift tırnak işaretlerini karıştırırsanız, sözdizimi hatası alırsınız.]
