## Introduction

Regression is where models predict a number. [Regresyon, modellerin bir sayıyı öngördüğü yerdir.]

In machine learning, the goal of regression is to create a model that can predict a numeric, quantifiable value, such as a price, amount, size, or other scalar number. [Makine öğreniminde regresyonun amacı, fiyat, miktar, boyut veya diğer skaler sayılar gibi sayısal, ölçülebilir bir değeri tahmin edebilen bir model oluşturmaktır.]

Regression is a statistical technique of fundamental importance to science because of its ease of interpretation, robustness, and speed in calculation. [Regresyon, yorumlama kolaylığı, sağlamlığı ve hesaplama hızı nedeniyle bilim için temel öneme sahip istatistiksel bir tekniktir.] Regression models provide an excellent foundation to understanding how more complex machine learning techniques work. [Regresyon modelleri, daha karmaşık makine öğrenimi tekniklerinin nasıl çalıştığını anlamak için mükemmel bir temel sağlar.]

In real world situations, particularly when little data are available, regression models are very useful for making predictions. [Gerçek dünya durumlarında, özellikle çok az veri mevcut olduğunda, regresyon modelleri tahmin yapmak için çok faydalıdır.] For example, if a company that rents bicycles wants to predict the expected number of rentals on a given day in the future, a regression model can predict this number. [Örneğin, bisiklet kiralayan bir şirket gelecekte belirli bir günde beklenen kiralama sayısını tahmin etmek isterse, bir regresyon modeli bu sayıyı tahmin edebilir.] A model could be created using existing data such as the number of bicycles that were rented on days where the season, day of the week, and so on, were also recorded. [Mevsim, haftanın günü vb. kayıtların tutulduğu günlerde kiralanan bisiklet sayısı gibi mevcut veriler kullanılarak bir model oluşturulabilir.]
