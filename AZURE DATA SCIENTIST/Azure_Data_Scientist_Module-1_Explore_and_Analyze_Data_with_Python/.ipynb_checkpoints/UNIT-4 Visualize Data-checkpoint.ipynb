## Visualize data

Data scientists visualize data to understand it better. [Veri bilimcileri, verileri daha iyi anlamak için görselleştirir.] This can mean looking at the raw data, summary measures such as averages, or graphing the data. [Bu, ham verilere bakmak, ortalamalar gibi özet ölçümler yapmak veya verilerin grafiğini çıkarmak anlamına gelebilir.] Graphs are a powerful means of viewing data, as we can discern moderately complex patterns quickly without needing to define mathematical summary measures. [Matematiksel özet ölçüleri tanımlamaya gerek kalmadan orta derecede karmaşık kalıpları hızlı bir şekilde ayırt edebildiğimiz için, grafikler verileri görüntülemenin güçlü bir yoludur.]

### Representing data visually

Representing data visually typically means graphing it. [Verileri görsel olarak temsil etmek, tipik olarak onu grafiklendirmek anlamına gelir.] This is done to provide a fast qualitative assessment of our data, which can be useful for understanding results, finding outlier values, understanding how numbers are distributed, and so on. [Bu, sonuçları anlamak, aykırı değerleri bulmak, sayıların nasıl dağıldığını anlamak vb. için yararlı olabilecek verilerimizin hızlı bir nitel değerlendirmesini sağlamak için yapılır.]

While sometimes we know ahead of time what kind of graph will be most useful, other times we use graphs in an exploratory way. [Bazen ne tür bir grafiğin en yararlı olacağını önceden bilirken, diğer zamanlarda grafikleri keşif amaçlı kullanırız.] To understand the power of data visualization, consider the data below: the location (x,y) of a self-driving car. [Veri görselleştirmenin gücünü anlamak için aşağıdaki verileri göz önünde bulundurun: sürücüsüz bir arabanın konumu (x,y).] In its raw form, it's hard to see any real patterns. [Ham haliyle, herhangi bir gerçek kalıp görmek zor.] The mean or average, tells us that its path was centred around x=0.2 and y=0.3, and the range of numbers appears to be between about -2 and 2. [Ortalama veya ortalama, yolunun x\u003d0.2 ve y\u003d0.3 etrafında ortalandığını ve sayı aralığının yaklaşık -2 ile 2 arasında göründüğünü söyler.]

If we now plot Location-X over time, we can see that we appear to have some missing values between times 7 and 12. [Şimdi Konum-X'i zaman içinde çizersek, 7 ile 12 saatleri arasında bazı eksik değerlere sahip olduğumuzu görebiliriz.]

If we graph X vs Y, we end up with a map of where the car has driven. [X'e karşı Y grafiğini çizersek, arabanın nereye gittiğini gösteren bir harita elde ederiz.] It’s instantly obvious that the car has been driving in a circle, but at some point drove to the center of that circle. [Arabanın bir daire çizdiği hemen belli oluyor, ama bir noktada bu dairenin merkezine gitti.]

Graphs aren't limited to 2D scatter plots like those above, but can be used to explore other kinds of data, like proportions - shown through pie charts, stacked bar graphs - how data are spread - with histograms, box and whisker plots - and how two data sets differ. [Grafikler, yukarıdakiler gibi 2B dağılım grafikleriyle sınırlı değildir, ancak diğer veri türlerini keşfetmek için kullanılabilir, örneğin pasta grafiklerle gösterilen oranlar, yığılmış çubuk grafikler - verilerin nasıl yayıldığı - histogramlar, kutu ve bıyık grafikleri ile - ve iki veri kümesinin nasıl farklı olduğu.] Often, when we're trying to understand raw data or results, we may experiment with different types of graphs until we come across one that explains the data in a visually intuitive way. [Çoğu zaman, ham verileri veya sonuçları anlamaya çalışırken, verileri görsel olarak sezgisel bir şekilde açıklayan bir grafikle karşılaşana kadar farklı grafik türlerini deneyebiliriz.]
