```python
import pandas as pd

# load the training dataset
# !wget https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/seeds.csv
data = pd.read_csv('seeds.csv')

# Display a random sample of 10 observations (just the features)
features = data[data.columns[0:6]]
features.sample(10)

from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

# Normalize the numeric features so they're on the same scale
scaled_features = MinMaxScaler().fit_transform(features[data.columns[0:6]])

# Get two principal components
pca = PCA(n_components=2).fit(scaled_features)
features_2d = pca.transform(scaled_features)
features_2d[0:10]
```

### K-Means Clustering

The algorithm we used to create our test clusters is K-Means. [Test kümelerimizi oluşturmak için kullandığımız algoritma K-Means'tir.] This is a commonly used clustering algorithm that separates a dataset into K clusters of equal variance. [Bu, bir veri kümesini eşit varyanslı K kümelerine ayıran yaygın olarak kullanılan bir kümeleme algoritmasıdır.] The number of clusters, K, is user defined. [Küme sayısı, K, kullanıcı tanımlıdır.] The basic algorithm has the following steps: [Temel algoritma aşağıdaki adımlara sahiptir:]

1. A set of K centroids are randomly chosen. [Rastgele bir K centroid kümesi seçilir.]

2. Clusters are formed by assigning the data points to their closest centroid. [Kümeler, veri noktalarının en yakın merkezlerine atanmasıyla oluşturulur.]

3. The means of each cluster is computed and the centroid is moved to the mean. [Her kümenin ortalamaları hesaplanır ve ağırlık merkezi ortalamaya taşınır.]

4. Steps 2 and 3 are repeated until a stopping criteria is met. [Adım 2 ve 3, bir durdurma kriteri karşılanana kadar tekrarlanır.] Typically, the algorithm terminates when each new iteration results in negligable movement of centroids and the clusters become static. [Tipik olarak, algoritma, her yeni yineleme, ağırlık merkezlerinin ihmal edilebilir hareketine yol açtığında ve kümeler statik hale geldiğinde sona erer.]

5. When the clusters stop changing, the algorithm has converged, defining the locations of the clusters - note that the random starting point for the centroids means that re-running the algorithm could result in slightly different clusters, so training usually involves multiple iterations, reinitializing the centroids each time, and the model with the best WCSS is selected. [Kümeler değişmeyi bıraktığında, algoritma birleşerek kümelerin konumlarını tanımlar - merkezler için rastgele başlangıç noktasının, algoritmayı yeniden çalıştırmanın biraz farklı kümelerle sonuçlanabileceği anlamına geldiğine dikkat edin, bu nedenle eğitim genellikle birden fazla yineleme içerir, her seferinde centroids ve en iyi WCSS'ye sahip model seçilir.]

Let's try using K-Means on our seeds data with a K value of 3. [K değeri 3 olan tohum verilerimizde K-Means kullanmayı deneyelim.]

```python
from sklearn.cluster import KMeans

# Create a model based on 3 centroids
model = KMeans(n_clusters=3, init='k-means++', n_init=100, max_iter=1000)
# Fit to the data and predict the cluster assignments for each data point
km_clusters = model.fit_predict(features.values)
# View the cluster assignments
km_clusters
```

Let's see those cluster assignments with the two-dimensional data points.

```python
import matplotlib.pyplot as plt
%matplotlib inline

def plot_clusters(samples, clusters):
    col_dic = {0:'blue',1:'green',2:'orange'}
    mrk_dic = {0:'*',1:'x',2:'+'}
    colors = [col_dic[x] for x in clusters]
    markers = [mrk_dic[x] for x in clusters]
    for sample in range(len(clusters)):
        plt.scatter(samples[sample][0], samples[sample][1], color = colors[sample], marker=markers[sample], s=100)
    plt.xlabel('Dimension 1')
    plt.ylabel('Dimension 2')
    plt.title('Assignments')
    plt.show()

plot_clusters(features_2d, km_clusters)
```

Hopefully, the the data has been separated into three distinct clusters. [Umarım, veriler üç ayrı kümeye ayrılmıştır.]

So what's the practical use of clustering? [Peki kümelemenin pratik kullanımı nedir?] In some cases, you may have data that you need to group into distict clusters without knowing how many clusters there are or what they indicate. [Bazı durumlarda, kaç tane küme olduğunu veya neyi gösterdiklerini bilmeden farklı kümeler halinde gruplamanız gereken verileriniz olabilir.] For example a marketing organization might want to separate customers into distinct segments, and then investigate how those segments exhibit different purchasing behaviors. [Örneğin bir pazarlama kuruluşu, müşterileri farklı segmentlere ayırmak ve ardından bu segmentlerin nasıl farklı satın alma davranışları sergilediğini araştırmak isteyebilir.]

Sometimes, clustering is used as an initial step towards creating a classification model. [Bazen kümeleme, bir sınıflandırma modeli oluşturmaya yönelik ilk adım olarak kullanılır.] You start by identifying distinct groups of data points, and then assign class labels to those clusters. [Farklı veri noktası gruplarını tanımlayarak başlarsınız ve ardından bu kümelere sınıf etiketleri atarsınız.] You can then use this labelled data to train a classification model. [Daha sonra bu etiketli verileri bir sınıflandırma modelini eğitmek için kullanabilirsiniz.]

In the case of the seeds data, the different species of seed are already known and encoded as 0 (Kama), 1 (Rosa), or 2 (Canadian), so we can use these identifiers to compare the species classifications to the clusters identified by our unsupervised algorithm [Tohum verileri söz konusu olduğunda, farklı tohum türleri zaten bilinmektedir ve 0 (Kama), 1 (Rosa) veya 2 (Kanadalı) olarak kodlanmıştır, bu nedenle tür sınıflandırmalarını tanımlanan kümelerle karşılaştırmak için bu tanımlayıcıları kullanabiliriz. denetimsiz algoritmamız tarafından]

```python
seed_species = data[data.columns[7]]
plot_clusters(features_2d, seed_species.values)
```

There may be some differences between the cluster assignments and class labels, but the K-Means model should have done a reasonable job of clustering the observations so that seeds of the same species are generally in the same cluster. [Küme atamaları ve sınıf etiketleri arasında bazı farklılıklar olabilir, ancak K-Means modeli, aynı türün tohumlarının genellikle aynı kümede olması için gözlemleri kümeleme konusunda makul bir iş çıkarmış olmalıdır.]

### Hierarchical Clustering

Hierarchical clustering methods make fewer distributional assumptions when compared to K-means methods. [Hiyerarşik kümeleme yöntemleri, K-ortalama yöntemleriyle karşılaştırıldığında daha az dağıtım varsayımı yapar.] However, K-means methods are generally more scalable, sometimes very much so. [Bununla birlikte, K-ortalama yöntemleri genellikle daha ölçeklenebilir, bazen çok daha fazla.]

Hierarchical clustering creates clusters by either a divisive method or agglomerative method. [Hiyerarşik kümeleme, kümeleri bölücü bir yöntemle veya aglomeratif yöntemle oluşturur.] The divisive method is a "top down" approach starting with the entire dataset and then finding partitions in a stepwise manner. [Bölücü yöntem, tüm veri kümesiyle başlayan ve daha sonra kademeli olarak bölümleri bulma 'yukarıdan aşağıya' bir yaklaşımdır.] Agglomerative clustering is a "bottom up** approach. [Aglomeratif kümeleme, 'aşağıdan yukarıya** bir yaklaşımdır.] In this lab you will work with agglomerative clustering which roughly works as follows: [Bu laboratuvarda kabaca aşağıdaki gibi çalışan aglomeratif kümeleme ile çalışacaksınız:]

1. The linkage distances between each of the data points is computed. [Veri noktalarının her biri arasındaki bağlantı mesafeleri hesaplanır.]

2. Points are clustered pairwise with their nearest neighbor. [Noktalar, en yakın komşularıyla ikili olarak kümelenir.]

3. Linkage distances between the clusters are computed. [Kümeler arasındaki bağlantı mesafeleri hesaplanır.]

4. Clusters are combined pairwise into larger clusters. [Kümeler çiftler halinde daha büyük kümeler halinde birleştirilir.]

5. Steps 3 and 4 are repeated until all data points are in a single cluster. [Tüm veri noktaları tek bir kümede olana kadar 3. ve 4. adımlar tekrarlanır.]

The linkage function can be computed in a number of ways: [Bağlantı fonksiyonu birkaç yolla hesaplanabilir:]

- Ward linkage measures the increase in variance for the clusters being linked, [Ward bağlantısı, bağlanan kümeler için varyanstaki artışı ölçer,]

- Average linkage uses the mean pairwise distance between the members of the two clusters, [Ortalama bağlantı, iki kümenin üyeleri arasındaki ortalama ikili mesafeyi kullanır,]

- Complete or Maximal linkage uses the maximum distance between the members of the two clusters. [Tam veya Maksimal bağlantı, iki kümenin üyeleri arasındaki maksimum mesafeyi kullanır.]

Several different distance metrics are used to compute linkage functions: [Bağlantı fonksiyonlarını hesaplamak için birkaç farklı mesafe metriği kullanılır:]

- Euclidian or l2 distance is the most widely used. [Öklid veya l2 mesafesi en yaygın kullanılanıdır.] This metric is only choice for the Ward linkage method. [Bu metrik, Ward bağlantı yöntemi için yalnızca seçimdir.]

- Manhattan or l1 distance is robust to outliers and has other interesting properties. [Manhattan veya l1 mesafesi aykırı değerlere karşı dayanıklıdır ve başka ilginç özelliklere sahiptir.]

- Cosine similarity, is the dot product between the location vectors divided by the magnitudes of the vectors. [Kosinüs benzerliği, konum vektörleri arasındaki nokta çarpımının vektörlerin büyüklüklerine bölünmesidir.] Notice that this metric is a measure of similarity, whereas the other two metrics are measures of difference. [Bu metriğin bir benzerlik ölçüsü olduğuna dikkat edin, diğer iki metriğin ise farklılık ölçüleri olduğuna dikkat edin.] Similarity can be quite useful when working with data such as images or text documents. [Benzerlik, resimler veya metin belgeleri gibi verilerle çalışırken oldukça faydalı olabilir.]

### Agglomerative Clustering

Let's see an example of clustering the seeds data using an agglomerative clustering algorithm.

```python
from sklearn.cluster import AgglomerativeClustering

agg_model = AgglomerativeClustering(n_clusters=3)
agg_clusters = agg_model.fit_predict(features.values)
agg_clusters
```

So what do the agglomerative cluster assignments look like?

```python
import matplotlib.pyplot as plt

%matplotlib inline

def plot_clusters(samples, clusters):
    col_dic = {0:'blue',1:'green',2:'orange'}
    mrk_dic = {0:'*',1:'x',2:'+'}
    colors = [col_dic[x] for x in clusters]
    markers = [mrk_dic[x] for x in clusters]
    for sample in range(len(clusters)):
        plt.scatter(samples[sample][0], samples[sample][1], color = colors[sample], marker=markers[sample], s=100)
    plt.xlabel('Dimension 1')
    plt.ylabel('Dimension 2')
    plt.title('Assignments')
    plt.show()

plot_clusters(features_2d, agg_clusters)
```
