## Evaluate different types of clustering

### Training a clustering model [Kümeleme modeli eğitimi]

There are multiple algorithms you can use for clustering. [Kümeleme için kullanabileceğiniz birden çok algoritma vardır.] One of the most commonly used algorithms is K-Means clustering that, in its simplest form, consists of the following steps: [En yaygın kullanılan algoritmalardan biri, en basit haliyle aşağıdaki adımlardan oluşan K-Means kümelemesidir:]
1. The feature values are vectorized to define n-dimensional coordinates (where n is the number of features). [Özellik değerleri, n boyutlu koordinatları tanımlamak için vektörleştirilir (burada n, özelliklerin sayısıdır).] In the flower example, we have two features (number of petals and number of leaves), so the feature vector has two coordinates that we can use to conceptually plot the data points in two-dimensional space. [Çiçek örneğinde iki özelliğimiz var (yaprak sayısı ve yaprak sayısı), bu nedenle özellik vektörünün iki boyutlu uzayda veri noktalarını kavramsal olarak çizmek için kullanabileceğimiz iki koordinatı var.]

2. You decide how many clusters you want to use to group the flowers, and call this value k. [Çiçekleri gruplamak için kaç tane küme kullanmak istediğinize siz karar verirsiniz ve bu değeri k olarak adlandırırsınız.] For example, to create three clusters, you would use a k value of 3. [Örneğin, üç küme oluşturmak için k değerini 3 kullanırsınız.] Then k points are plotted at random coordinates. [Daha sonra rastgele koordinatlarda k nokta çizilir.] These points will ultimately be the center points for each cluster, so they're referred to as centroids. [Bu noktalar nihayetinde her kümenin merkez noktaları olacaktır, bu nedenle merkez noktaları olarak anılırlar.]

3. Each data point (in this case flower) is assigned to its nearest centroid. [Her veri noktası (bu durumda çiçek) en yakın merkez noktasına atanır.]

4. Each centroid is moved to the center of the data points assigned to it based on the mean distance between the points. [Her bir centroid, noktalar arasındaki ortalama mesafeye göre kendisine atanan veri noktalarının merkezine taşınır.]

5. After moving the centroid, the data points may now be closer to a different centroid, so the data points are reassigned to clusters based on the new closest centroid. [Centroid hareket ettirildikten sonra, veri noktaları artık farklı bir centroide daha yakın olabilir, bu nedenle veri noktaları yeni en yakın centroide dayalı olarak kümelere yeniden atanır.]

6. The centroid movement and cluster reallocation steps are repeated until the clusters become stable or a pre-determined maximum number of iterations is reached. [Merkez hareketi ve küme yeniden tahsis adımları, kümeler kararlı hale gelene veya önceden belirlenmiş bir maksimum yineleme sayısına ulaşılana kadar tekrarlanır.]

### Hierarchical Clustering

Hierarchical clustering is another type of clustering algorithm in which clusters themselves belong to a larger group, which belong to even larger groups, and so on. [Hiyerarşik kümeleme, kümelerin kendilerinin daha büyük bir gruba, hatta daha büyük gruplara ait olduğu vb. başka bir kümeleme algoritması türüdür.] The result is that data points can be clusters in differing degrees of precision: with a large number of very small and precise groups, or a small number of larger groups. [Sonuç olarak, veri noktaları farklı kesinlik derecelerinde kümeler olabilir: çok sayıda çok küçük ve kesin grup veya az sayıda daha büyük grup.]

For example, if we apply clustering to the meanings of words, we may get a group containing adjectives specific to emotions (‘angry’, ‘happy’, and so on), which itself belongs to a group containing all human-related adjectives (‘happy’, ‘handsome’, ‘young’), and this belongs to an even higher group containing all adjectives (‘happy’, ‘green’, ‘handsome’, ‘hard’ etc.). [Örneğin, kelimelerin anlamlarına kümeleme uygularsak, kendisi de insanla ilgili tüm sıfatları içeren bir gruba ait olan duygulara özgü sıfatları ('kızgın', 'mutlu' vb.) içeren bir grup elde edebiliriz. 'mutlu', 'yakışıklı', 'genç') ve bu, tüm sıfatları ('mutlu', 'yeşil', 'yakışıklı', 'zor' vb.) içeren daha yüksek bir gruba aittir.]

Hierarchical clustering is useful for not only breaking data into groups, but understanding the relationships between these groups. [Hiyerarşik kümeleme, yalnızca verileri gruplara ayırmak için değil, aynı zamanda bu gruplar arasındaki ilişkileri anlamak için de yararlıdır.] A major advantage of hierarchical clustering is that it does not require the number of clusters to be defined in advance, and can sometimes provide more interpretable results than non-hierarchical approaches. [Hiyerarşik kümelemenin önemli bir avantajı, önceden tanımlanacak küme sayısının gerektirmemesi ve bazen hiyerarşik olmayan yaklaşımlardan daha fazla yorumlanabilir sonuçlar sağlayabilmesidir.] The major drawback is that these approaches can take much longer to compute than simpler approaches and sometimes are not suitable for large datasets. [En büyük dezavantajı, bu yaklaşımların hesaplanmasının daha basit yaklaşımlardan çok daha uzun sürebilmesi ve bazen büyük veri kümeleri için uygun olmamasıdır.]
