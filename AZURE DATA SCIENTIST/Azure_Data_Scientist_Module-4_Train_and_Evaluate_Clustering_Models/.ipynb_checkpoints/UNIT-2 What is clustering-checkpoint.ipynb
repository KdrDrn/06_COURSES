## What is clustering?

Clustering is a form of unsupervised machine learning in which observations are grouped into clusters based on similarities in their data values, or features. [Kümeleme, gözlemlerin veri değerlerindeki veya özelliklerindeki benzerliklere göre kümeler halinde gruplandırıldığı bir denetimsiz makine öğrenimi biçimidir.] This kind of machine learning is considered unsupervised because it does not make use of previously known label values to train a model; in a clustering model, the label is the cluster to which the observation is assigned, based purely on its features. [Bu tür bir makine öğrenimi, bir modeli eğitmek için önceden bilinen etiket değerlerini kullanmadığı için denetimsiz olarak kabul edilir; Bir kümeleme modelinde etiket, yalnızca özelliklerine dayalı olarak, gözlemin atandığı kümedir.]

For example, suppose a botanist observes a sample of flowers and records the number of petals and leaves on each flower. [Örneğin, bir botanikçinin bir çiçek örneğini gözlemlediğini ve her bir çiçeğin üzerindeki taçyaprak ve yaprak sayısını kaydettiğini varsayalım.]

It may be useful to group these flowers into clusters based on similarities between their features. [Bu çiçekleri özellikleri arasındaki benzerliklere göre kümeler halinde gruplandırmak faydalı olabilir.]

There are many ways this could be done. [Bunu yapmanın birçok yolu var.] For example, if most flowers have the same number of leaves, they could be grouped into those with many vs few petals. [Örneğin, çoğu çiçek aynı sayıda yaprağa sahipse, çok veya az taç yaprağı olanlar şeklinde gruplandırılabilirler.] Alternatively, if both petal and leaf counts vary considerably there may be a pattern to discover, such as those with many leaves also having many petals. [Alternatif olarak, eğer hem petal hem de yaprak sayıları önemli ölçüde farklılık gösteriyorsa, örneğin çok sayıda yaprağı olan ve aynı zamanda birçok taç yaprağına sahip olanlar gibi keşfedilecek bir model olabilir.] The goal of the clustering algorithm is to find the optimal way to split the dataset into groups. [Kümeleme algoritmasının amacı, veri kümesini gruplara ayırmanın en uygun yolunu bulmaktır.] What ‘optimal’ means depends on both the algorithm used and the dataset that is provided. ['Optimal'in ne anlama geldiği, hem kullanılan algoritmaya hem de sağlanan veri kümesine bağlıdır.]

Although this flower example may be simple for a human to achieve with only a few samples, as the dataset grows to thousands of samples or to more than two features, clustering algorithms become very useful to quickly dissect a dataset into groups. [Bu çiçek örneğinin, bir insan için yalnızca birkaç örnekle elde etmesi basit olsa da, veri kümesi binlerce örneğe veya ikiden fazla özelliğe büyüdükçe, kümeleme algoritmaları, bir veri kümesini hızlı bir şekilde gruplara ayırmak için çok kullanışlı hale gelir.]
