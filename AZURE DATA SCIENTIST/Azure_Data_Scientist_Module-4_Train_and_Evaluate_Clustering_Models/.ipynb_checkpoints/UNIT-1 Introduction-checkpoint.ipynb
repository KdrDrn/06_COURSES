## Introduction

Clustering is the process of grouping objects with similar objects. For example, in the image below we have a collection of 2D coordinates that have been clustered into three categories - top left (yellow), bottom (red), and top right (blue).

A major difference between clustering and classification models is that clustering is an ‘unsupervised’ method, where ‘training’ is done without labels. [Kümeleme ve sınıflandırma modelleri arasındaki önemli bir fark, kümelemenin 'eğitimin' etiketler olmadan yapıldığı 'denetimsiz' bir yöntem olmasıdır.] Instead, models identify examples that have a similar collection of features. [Bunun yerine modeller, benzer özellikler koleksiyonuna sahip örnekleri tanımlar.] In the image above, examples that are in a similar location are grouped together. [Yukarıdaki resimde benzer konumdaki örnekler bir arada gruplandırılmıştır.]

Clustering is common and useful for exploring new data where patterns between data points, such as high-level categories, are not yet known. [Kümeleme, üst düzey kategoriler gibi veri noktaları arasındaki kalıpların henüz bilinmediği yeni verileri keşfetmek için yaygın ve kullanışlıdır.] It's used in many fields that need to automatically label complex data, including analysis of social networks, brain connectivity, spam filtering, and so on. [Sosyal ağların analizi, beyin bağlantısı, spam filtreleme vb. dahil olmak üzere karmaşık verileri otomatik olarak etiketlemesi gereken birçok alanda kullanılır.]
