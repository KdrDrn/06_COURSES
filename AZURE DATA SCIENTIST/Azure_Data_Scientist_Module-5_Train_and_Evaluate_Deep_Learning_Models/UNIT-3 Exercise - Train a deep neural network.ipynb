## Exercise - Train a deep neural network

So far in this module, you've learned a lot about the theory and principles of deep learning with neural networks. [Şimdiye kadar bu modülde, sinir ağlarıyla derin öğrenmenin teorisi ve ilkeleri hakkında çok şey öğrendiniz.] The best way to learn how to apply this theory is to actually build a deep learning model, and that's what you'll do in this exercise. [Bu teoriyi nasıl uygulayacağınızı öğrenmenin en iyi yolu, aslında bir derin öğrenme modeli oluşturmaktır ve bu alıştırmada bunu yapacaksınız.]

There are many frameworks available for training deep neural networks, and in this exercise you can choose to explore either (or both) of two of the most popular deep learning frameworks for Python: PyTorch and TensorFlow. [Derin sinir ağlarını eğitmek için kullanılabilecek birçok çerçeve vardır ve bu alıştırmada Python için en popüler derin öğrenme çerçevelerinden ikisini (veya her ikisini) keşfetmeyi seçebilirsiniz: PyTorch ve TensorFlow.]

### Create an Azure Machine Learning workspace

If you don't already have an Azure Machine Learning workspace in your Azure subscription, follow these steps to create one:

1. Sign into the Azure portal using the Microsoft account associated with your Azure subscription.

2. On the Azure Home page, under Azure services, select Create a resource. The Create a resource pane appears.

3. In the Search services and marketplace search box, search for and select Machine Learning. The Machine Learning pane appears.

4. Select Create. The Machine learning pane appears.

5. On the Basics tab, Enter the following values to each setting.



6. Accept the remaining defaults, and select Review + create.

7. After validation passes, select Create.

   Wait for your workspace resource to be created as it can take a few minutes.

8. When deployment completes, select Go to resource. Your Machine learning pane appears.

9. Select Launch studio, or go to https://ml.azure.com, and sign in using your Microsoft account. The Microsoft Azure Machine Learning Studio page appears.

10. In Azure Machine Learning Studio, toggle the ☰ icon at the top left to expane/collapse its menu pane. You can use these options to manage the resources in your workspace.

### Create a compute instance

To run the notebook used in this exercise, you will need a compute instance in your Azure Machine Learning workspace. [Bu alıştırmada kullanılan not defterini çalıştırmak için Azure Machine Learning çalışma alanınızda bir işlem örneğine ihtiyacınız olacak.]

1. In the left menu pane, under Manage, select Compute. [Sol menü bölmesinde, Yönet altında Hesapla'yı seçin.] The Compute pane appears. [Hesaplama bölmesi görünür.]

2. On the Compute Instances tab, if you already have a compute instance, start it; otherwise, create a new compute instance by selecting New. [İşlem Eşgörünümleri sekmesinde, zaten bir işlem bulut sunucunuz varsa onu başlatın; aksi takdirde, Yeni'yi seçerek yeni bir işlem örneği oluşturun.] The Create compute instance pane appears. [İşlem örneği oluştur bölmesi görünür.]

3. Enter the following values for each setting: [Her ayar için aşağıdaki değerleri girin:]

- Compute name: enter a unique name [Hesaplama adı: benzersiz bir ad girin]

- Virtual machine type: CPU [Sanal makine türü: CPU]

- Virtual machine size: Select from recommended options: Standard_DS11_v2 [Sanal makine boyutu: Önerilen seçenekler arasından 
seçim yapın: Standard_DS11_v2]

4. Select Create. [Oluştur'u seçin.] The Compute pane reappears with your Compute instance listed. [Hesaplama bölmesi, Hesaplama örneğiniz listelenmiş olarak yeniden görünür.]

5. Wait for the compute instance to start as this may take a couple of minutes. [Bu işlem birkaç dakika sürebileceğinden işlem örneğinin başlamasını bekleyin.] Under the State column, your Compute instance will change to Running. [Durum sütununun altında, Hesaplama örneğiniz Çalışıyor olarak değişecektir.]
