## Deep neural network concepts

Before exploring how to train a deep neural network (DNN) machine learning model, let's consider what we're trying to achieve. [Derin sinir a (DNN) makine renimi modelinin nasl eitileceini kefetmeden nce, neyi baarmaya altmz dnelim.] Machine learning is concerned with predicting a label based on some features of a particular observation. [Makine renimi, belirli bir gzlemin baz zelliklerine dayal olarak bir etiketin tahmin edilmesiyle ilgilenir.] In simple terms, a machine learning model is a function that calculates [Basit bir ifadeyle, bir makine renimi modeli, hesaplayan bir fonksiyondur.] y (the label) from x (the features): f(x)=y.

### A deep neural network model

So how would we use deep learning to build a classification model for the penguin classification model? [Peki penguen snflandrma modeli iin bir snflandrma modeli oluturmak iin derin renmeyi nasl kullanrz?] Let's look at an example: [Bir rnee bakalm:]



The deep neural network model for the classifier consists of multiple layers of artificial neurons. [Snflandrc iin derin sinir a modeli, oklu yapay nron katmanlarndan oluur.] In this case, there are four layers: [Bu durumda, drt katman vardr:]

- An input layer with a neuron for each expected input (x) value. [Beklenen her giri (x) deeri iin bir nron ieren bir giri katman.]

- Two so-called hidden layers, each containing five neurons. [Her biri be nron ieren iki gizli katman.]

- An output layer containing three neurons - one for each class probability (y) value to be predicted by the model. [Model tarafndan tahmin edilecek her snf olaslk (y) deeri iin bir tane olmak zere  nron ieren bir kt katman.]

Because of the layered architecture of the network, this kind of model is sometimes referred to as a multilayer perceptron. [An katmanl mimarisi nedeniyle, bu tr bir modele bazen ok katmanl alglayc denir.] Additionally, notice that all neurons in the input and hidden layers are connected to all neurons in the subsequent layers - this is an example of a fully connected network. [Ek olarak, giri ve gizli katmanlardaki tm nronlarn sonraki katmanlardaki tm nronlara bal olduuna dikkat edin - bu tam balantl bir aa bir rnektir.]

When you create a model like this, you must define an input layer that supports the number of features your model will process, and an output layer that reflects the number of outputs you expect it to produce. [Bunun gibi bir model oluturduunuzda, modelinizin ileyecei zellik saysn destekleyen bir girdi katman ve retmesini beklediiniz kt saysn yanstan bir kt katman tanmlamanz gerekir.] You can decide how many hidden layers you want to include and how many neurons are in each of them; but you have no control over the input and output values for these layers - these are determined by the model training process. [Ka tane gizli katman eklemek istediinize ve bunlarn her birinde ka nron olduuna karar verebilirsiniz; ancak bu katmanlar iin giri ve k deerleri zerinde hibir kontrolnz yoktur - bunlar model eitim sreci tarafndan belirlenir.]

### Training a deep neural network

The training process for a deep neural network consists of multiple iterations, called epochs. [Derin bir sinir a iin eitim sreci, dnemler ad verilen oklu yinelemelerden oluur.] For the first epoch, you start by assigning random initialization values for the weight (w) and bias b values. [lk dnem iin arlk (w) ve sapma b deerleri iin rastgele balatma deerleri atayarak balarsnz.] Then the process is as follows: [Daha sonra sre u ekildedir:]

1. Features for data observations with known label values are submitted to the input layer. [Bilinen etiket deerlerine sahip veri gzlemleri iin zellikler girdi katmanna gnderilir.] Generally, these observations are grouped into batches (often referred to as mini-batches). [Genel olarak, bu gzlemler gruplar halinde gruplandrlr (genellikle mini partiler olarak adlandrlr).]

2. The neurons then apply their function, and if activated, pass the result onto the next layer until the output layer produces a prediction. [Nronlar daha sonra ilevlerini uygular ve eer etkinletirilirse, kt katman bir tahmin retene kadar sonucu bir sonraki katmana iletir.]

3. The prediction is compared to the actual known value, and the amount of variance between the predicted and true values (which we call the loss) is calculated. [Tahmin, bilinen gerek deerle karlatrlr ve tahmin edilen ve gerek deerler (kayp dediimiz) arasndaki varyans miktar hesaplanr.]

4. Based on the results, revised values for the weights and bias values are calculated to reduce the loss, and these adjustments are backpropagated to the neurons in the network layers. [Sonulara dayal olarak, arlklar ve bias deerleri iin revize edilmi deerler, kayb azaltmak iin hesaplanmakta ve bu ayarlamalar a katmanlarndaki nronlara geri yaylmaktadr.]

5. The next epoch repeats the batch training forward pass with the revised weight and bias values, hopefully improving the accuracy of the model (by reducing the loss). [Bir sonraki dnem, gzden geirilmi arlk ve sapma deerleriyle toplu eitim ileri geiini tekrarlar ve umarz modelin doruluunu artrr (kayb azaltarak).]

**Note**

Processing the training features as a batch improves the efficiency of the training process by processing multiple observations simultaneously as a matrix of features with vectors of weights and biases. [Eitim zelliklerini toplu olarak ilemek, birden ok gzlemi arlk ve nyarg vektrleriyle bir zellik matrisi olarak ayn anda ileyerek eitim srecinin verimliliini artrr.] Linear algebraic functions that operate with matrices and vectors also feature in 3D graphics processing, which is why computers with graphic processing units (GPUs) provide significantly better performance for deep learning model training than central processing unit (CPU) only computers. [Matrisler ve vektrlerle alan dorusal cebirsel ilevler, 3B grafik ilemede de yer alr; bu nedenle, grafik ileme birimlerine (GPU'lar) sahip bilgisayarlar, derin renme modeli eitimi iin yalnzca merkezi ilem birimi (CPU) olan bilgisayarlardan nemli lde daha iyi performans salar.]

### A closer look at loss functions and backpropagation

The previous description of the deep learning training process mentioned that the loss from the model is calculated and used to adjust the weight and bias values. [Derin renme eitim srecinin nceki aklamas, modeldeki kaybn hesaplandndan ve arlk ve yanllk deerlerini ayarlamak iin kullanldndan bahsetmiti.] How exactly does this work? [Bu tam olarak nasl alyor?]

### Calculating loss

Suppose one of the samples passed through the training process contains features of an Adelie specimen (class 0). [Eitim srecinden geen rneklerden birinin bir Adelie rneinin (snf 0) zelliklerini ierdiini varsayalm.] The correct output from the network would be [1, 0, 0]. [Adan doru kt [1, 0, 0] olacaktr.] Now suppose that the output produced by the network is [0.4, 0.3, 0.3]. [imdi, a tarafndan retilen ktnn [0.4, 0.3, 0.3] olduunu varsayalm.] Comparing these, we can calculate an absolute variance for each element (in other words, how far is each predicted value away from what it should be) as [0.6, 0.3, 0.3]. [Bunlar karlatrarak, her eleman iin mutlak bir varyans (dier bir deyile, tahmin edilen her deerin olmas gerekenden ne kadar uzakta olduunu) [0.6, 0.3, 0.3] olarak hesaplayabiliriz.]

In reality, since we're actually dealing with multiple observations, we typically aggregate the variance - for example by squaring the individual variance values and calculating the mean, so we end up with a single, average loss value, like 0.18. [Gerekte, aslnda birden ok gzlemle uratmz iin, tipik olarak varyans topluyoruz - rnein bireysel varyans deerlerinin karesini alarak ve ortalamay hesaplayarak, bylece 0,18 gibi tek bir ortalama kayp deeri elde ederiz.]

### Optimizers

Now, here's the clever bit. [imdi, ite zekice olan ksm.] The loss is calculated using a function, which operates on the results from the final layer of the network, which is also a function. [Kayp, ayn zamanda bir fonksiyon olan an son katmanndan elde edilen sonular zerinde alan bir fonksiyon kullanlarak hesaplanr.] The final layer of network operates on the outputs from the previous layers, which are also functions. [An son katman, ayn zamanda ilevler olan nceki katmanlarn ktlar zerinde alr.] So in effect, the entire model from the input layer right through to the loss calculation is just one big nested function. [Yani aslnda, girdi katmanndan kayp hesaplamasna kadar tm model yalnzca byk bir i ie ilevdir.] Functions have a few really useful characteristics, including: [levler, aadakiler de dahil olmak zere gerekten yararl birka zellie sahiptir:]

- You can conceptualize a function as a plotted line comparing its output with each of its variables. [Bir ilevi, ktsn deikenlerinin her biriyle karlatran izilmi bir izgi olarak kavramsallatrabilirsiniz.]

- You can use differential calculus to calculate the derivative of the function at any point with respect to its variables. [Deikenlere gre herhangi bir noktada fonksiyonun trevini hesaplamak iin diferansiyel hesab kullanabilirsiniz.]

Let's take the first of these capabilities. [Bu yeteneklerden ilkini ele alalm.] We can plot the line of the function to show how an individual weight value compares to loss, and mark on that line the point where the current weight value matches the current loss value. [Bireysel arlk deerinin kaypla nasl karlatrldn gstermek iin fonksiyonun satrn izebiliriz ve bu satrda mevcut arlk deerinin mevcut kayp deeriyle eletii noktay iaretleyebiliriz.]



Now let's apply the second characteristic of a function. [imdi bir fonksiyonun ikinci zelliini uygulayalm.] The derivative of a function for a given point indicates whether the slope (or gradient) of the function output (in this case, loss) is increasing or decreasing with respect to a function variable (in this case, the weight value). [Belirli bir nokta iin bir fonksiyonun trevi, fonksiyon ktsnn (bu durumda kayp) eiminin (veya gradyannn) bir fonksiyon deikenine (bu durumda arlk deeri) gre artan m yoksa azalan m olduunu gsterir.] A positive derivative indicates that the function is increasing, and a negative derivative indicates that it is decreasing. [Pozitif trev, fonksiyonun arttn, negatif trev ise azalmakta olduunu gsterir.] In this case, at the plotted point for the current weight value, the function has a downward gradient. [Bu durumda, mevcut arlk deeri iin izilen noktada, fonksiyon aa doru bir eime sahiptir.] In other words, increasing the weight will have the effect of decreasing the loss. [Baka bir deyile, arl artrmak, kayb azaltc etkiye sahip olacaktr.]

We use an optimizer to apply this same trick for all of the weight and bias variables in the model and determine in which direction we need to adjust them (up or down) to reduce the overall amount of loss in the model. [Ayn numaray modeldeki tm arlk ve sapma deikenlerine uygulamak iin bir optimize edici kullanrz ve modeldeki toplam kayp miktarn azaltmak iin bunlar hangi ynde (yukar veya aa) ayarlamamz gerektiini belirleriz.] There are multiple commonly used optimization algorithms, including stochastic gradient descent (SGD), Adaptive Learning Rate (ADADELTA), Adaptive Momentum Estimation (Adam), and others; all of which are designed to figure out how to adjust the weights and biases to minimize loss. [Stokastik gradyan inii (SGD), Uyarlanabilir renme Hz (ADADELTA), Uyarlanabilir Momentum Tahmini (Adam) ve dierleri dahil olmak zere yaygn olarak kullanlan birden ok optimizasyon algoritmas vardr; bunlarn tm, kayb en aza indirmek iin arlklarn ve nyarglarn nasl ayarlanacan bulmak iin tasarlanmtr.]

### Learning rate

Now, the obvious next question is, by how much should the optimizer adjust the weights and bias values? [imdi, bir sonraki bariz soru, optimize edicinin arlklar ve sapma deerlerini ne kadar ayarlamas gerektiidir?] If you look at the plot for our weight value, you can see that increasing the weight by a small amount will follow the function line down (reducing the loss), but if we increase it by too much, the function line starts to go up again, so we might actually increase the loss; and after the next epoch, we might find we need to reduce the weight. [Arlk deerimizin grafiine bakarsanz, arl kk bir miktar artrmann fonksiyon satrn takip edeceini (kayb azaltarak) grebilirsiniz, ancak ok fazla arttrrsak fonksiyon satr yukar kmaya balar. tekrar, bylece aslnda kayb artrabiliriz; ve bir sonraki adan sonra, arl azaltmamz gerektiini grebiliriz.]
The size of the adjustment is controlled by a parameter that you set for training called the learning rate. [Ayarlamann boyutu, eitim iin belirlediiniz, renme oran ad verilen bir parametre tarafndan kontrol edilir.] A low learning rate results in small adjustments (so it can take more epochs to minimize the loss), while a high learning rate results in large adjustments (so you might miss the minimum altogether). [Dk bir renme oran kk ayarlamalarla sonulanr (bylece kayb en aza indirmek daha fazla dnem alabilir), yksek bir renme oran ise byk ayarlamalarla sonulanr (bylece minimumu tamamen karabilirsiniz).]
