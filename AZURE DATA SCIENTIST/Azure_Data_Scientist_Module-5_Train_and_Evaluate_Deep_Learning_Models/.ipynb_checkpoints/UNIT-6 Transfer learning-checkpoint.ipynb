## Transfer learning

In life, it’s often easier to learn a new skill if you already have expertise in a similar, transferrable skill. [Hayatta, benzer, aktarılabilir bir beceride zaten uzmanlığınız varsa, yeni bir beceri öğrenmek genellikle daha kolaydır.] For example, it’s probably easier to teach someone how to drive a bus if they have already learned how to drive a car. [Örneğin, araba kullanmayı zaten öğrenmiş olan birine otobüs kullanmayı öğretmek muhtemelen daha kolaydır.] The driver can build on the driving skills they've already learned in a car, and apply them to driving a bus. [Sürücü, bir arabada zaten öğrendikleri sürüş becerilerini geliştirebilir ve bunları bir otobüse uygulayabilir.]

The same principle can be applied to training deep learning models through a technique called transfer learning. [Aynı ilke, derin öğrenme modellerini transfer öğrenme adı verilen bir teknikle eğitmek için de uygulanabilir.]

### How transfer learning works
A Convolutional Neural Network (CNN) for image classification is typically composed of multiple layers that extract features, and then use a final fully connected layer to classify images based on these features. [Görüntü sınıflandırması için bir Evrişimsel Sinir Ağı (CNN), tipik olarak, özellikleri çıkaran ve ardından görüntüleri bu özelliklere dayalı olarak sınıflandırmak için son bir tam bağlantılı katman kullanan birden çok katmandan oluşur.]



Conceptually, this neural network consists of two distinct sets of layers: [Kavramsal olarak, bu sinir ağı iki farklı katman kümesinden oluşur:]

1. A set of layers from the base model that perform feature extraction. [Özellik çıkarma işlemini gerçekleştiren temel modelden bir katmanlar kümesi.]

2. A fully connected layer that takes the extracted features and uses them for class prediction. [Çıkarılan özellikleri alan ve bunları sınıf tahmini için kullanan tam bağlantılı bir katman.]

The feature extraction layers apply convolutional filters and pooling to emphasize edges, corners, and other patterns in the images that can be used to differentiate them, and in theory should work for any set of images with the same dimensions as the input layer of the network. [Özellik çıkarma katmanları, onları ayırt etmek için kullanılabilecek görüntülerdeki kenarları, köşeleri ve diğer desenleri vurgulamak için evrişimsel filtreler ve havuzlama uygular ve teorik olarak ağın giriş katmanıyla aynı boyutlara sahip herhangi bir görüntü kümesi için çalışmalıdır. .] The prediction layer maps the features to a set of outputs that represent probabilities for each class label you want to use to classify the images. [Tahmin katmanı, özellikleri, görüntüleri sınıflandırmak için kullanmak istediğiniz her sınıf etiketi için olasılıkları temsil eden bir dizi çıktıyla eşler.]

By separating the network into these types of layers, we can take the feature extraction layers from a model that has already been trained and append one or more layers to use the extracted features for prediction of the appropriate class labels for your images. [Ağı bu tür katmanlara ayırarak, daha önce eğitilmiş bir modelden özellik çıkarma katmanlarını alabilir ve görüntüleriniz için uygun sınıf etiketlerinin tahmini için çıkarılan özellikleri kullanmak üzere bir veya daha fazla katman ekleyebiliriz.] This approach enables you to keep the pre-trained weights for the feature extraction layers, which means you only need to train the prediction layers you have added. [Bu yaklaşım, özellik çıkarma katmanları için önceden eğitilmiş ağırlıkları korumanıza olanak tanır; bu, yalnızca eklediğiniz tahmin katmanlarını eğitmeniz gerektiği anlamına gelir.]

There are many established convolutional neural network architectures for image classification that you can use as the base model for transfer learning, so you can build on the work someone else has already done to easily create an effective image classification model. [Aktarım öğrenimi için temel model olarak kullanabileceğiniz görüntü sınıflandırması için birçok yerleşik evrişimli sinir ağı mimarisi vardır, böylece etkili bir görüntü sınıflandırma modelini kolayca oluşturmak için başka birinin zaten yapmış olduğu işi geliştirebilirsiniz.]
