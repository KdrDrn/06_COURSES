## Transfer learning

In life, its often easier to learn a new skill if you already have expertise in a similar, transferrable skill. [Hayatta, benzer, aktarlabilir bir beceride zaten uzmanlnz varsa, yeni bir beceri renmek genellikle daha kolaydr.] For example, its probably easier to teach someone how to drive a bus if they have already learned how to drive a car. [rnein, araba kullanmay zaten renmi olan birine otobs kullanmay retmek muhtemelen daha kolaydr.] The driver can build on the driving skills they've already learned in a car, and apply them to driving a bus. [Src, bir arabada zaten rendikleri sr becerilerini gelitirebilir ve bunlar bir otobse uygulayabilir.]

The same principle can be applied to training deep learning models through a technique called transfer learning. [Ayn ilke, derin renme modellerini transfer renme ad verilen bir teknikle eitmek iin de uygulanabilir.]

### How transfer learning works
A Convolutional Neural Network (CNN) for image classification is typically composed of multiple layers that extract features, and then use a final fully connected layer to classify images based on these features. [Grnt snflandrmas iin bir Evriimsel Sinir A (CNN), tipik olarak, zellikleri karan ve ardndan grntleri bu zelliklere dayal olarak snflandrmak iin son bir tam balantl katman kullanan birden ok katmandan oluur.]



Conceptually, this neural network consists of two distinct sets of layers: [Kavramsal olarak, bu sinir a iki farkl katman kmesinden oluur:]

1. A set of layers from the base model that perform feature extraction. [zellik karma ilemini gerekletiren temel modelden bir katmanlar kmesi.]

2. A fully connected layer that takes the extracted features and uses them for class prediction. [karlan zellikleri alan ve bunlar snf tahmini iin kullanan tam balantl bir katman.]

The feature extraction layers apply convolutional filters and pooling to emphasize edges, corners, and other patterns in the images that can be used to differentiate them, and in theory should work for any set of images with the same dimensions as the input layer of the network. [zellik karma katmanlar, onlar ayrt etmek iin kullanlabilecek grntlerdeki kenarlar, keleri ve dier desenleri vurgulamak iin evriimsel filtreler ve havuzlama uygular ve teorik olarak an giri katmanyla ayn boyutlara sahip herhangi bir grnt kmesi iin almaldr. .] The prediction layer maps the features to a set of outputs that represent probabilities for each class label you want to use to classify the images. [Tahmin katman, zellikleri, grntleri snflandrmak iin kullanmak istediiniz her snf etiketi iin olaslklar temsil eden bir dizi ktyla eler.]

By separating the network into these types of layers, we can take the feature extraction layers from a model that has already been trained and append one or more layers to use the extracted features for prediction of the appropriate class labels for your images. [A bu tr katmanlara ayrarak, daha nce eitilmi bir modelden zellik karma katmanlarn alabilir ve grntleriniz iin uygun snf etiketlerinin tahmini iin karlan zellikleri kullanmak zere bir veya daha fazla katman ekleyebiliriz.] This approach enables you to keep the pre-trained weights for the feature extraction layers, which means you only need to train the prediction layers you have added. [Bu yaklam, zellik karma katmanlar iin nceden eitilmi arlklar korumanza olanak tanr; bu, yalnzca eklediiniz tahmin katmanlarn eitmeniz gerektii anlamna gelir.]

There are many established convolutional neural network architectures for image classification that you can use as the base model for transfer learning, so you can build on the work someone else has already done to easily create an effective image classification model. [Aktarm renimi iin temel model olarak kullanabileceiniz grnt snflandrmas iin birok yerleik evriimli sinir a mimarisi vardr, bylece etkili bir grnt snflandrma modelini kolayca oluturmak iin baka birinin zaten yapm olduu ii gelitirebilirsiniz.]
