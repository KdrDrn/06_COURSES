## Introduction

Deep learning is an advanced form of machine learning that tries to emulate the way the human brain learns. [Derin öğrenme, insan beyninin öğrenme şeklini taklit etmeye çalışan gelişmiş bir makine öğrenimi biçimidir.]

In your brain, you have nerve cells called neurons, which are connected to one another by nerve extensions that pass electrochemical signals through the network. [Beyninizde, ağ üzerinden elektrokimyasal sinyaller ileten sinir uzantıları ile birbirine bağlanan nöron adı verilen sinir hücreleriniz vardır.]

When the first neuron in the network is stimulated, the input signal is processed, and if it exceeds a particular threshold, the neuron is activated and passes the signal on to the neurons to which it is connected. [Ağdaki ilk nöron uyarıldığında giriş sinyali işlenir ve belirli bir eşiği aşarsa nöron aktive olur ve sinyali bağlı olduğu nöronlara iletir.] These neurons in turn may be activated and pass the signal on through the rest of the network. [Bu nöronlar sırayla etkinleştirilebilir ve sinyali ağın geri kalanına iletebilir.] Over time, the connections between the neurons are strengthened by frequent use as you learn how to respond effectively. [Zamanla, etkin bir şekilde nasıl tepki vereceğinizi öğrendikçe, nöronlar arasındaki bağlantılar sık kullanımla güçlenir.] For example, if someone throws a ball towards you, your neuron connections enable you to process the visual information and coordinate your movements to catch the ball. [Örneğin, biri size bir top atarsa, nöron bağlantılarınız, görsel bilgileri işlemenizi ve topu yakalamak için hareketlerinizi koordine etmenizi sağlar.] If you perform this action repeatedly, the network of neurons involved in catching a ball will grow stronger as you learn how to be better at catching a ball. [Bu eylemi tekrar tekrar yaparsanız, bir topu yakalamada nasıl daha iyi olacağınızı öğrendikçe, bir topu yakalamaya dahil olan nöronlar ağı güçlenecektir.]

Deep learning emulates this biological process using artificial neural networks that process numeric inputs rather than electrochemical stimuli. [Derin öğrenme, elektrokimyasal uyaranlar yerine sayısal girdileri işleyen yapay sinir ağlarını kullanarak bu biyolojik süreci taklit eder.]

The incoming nerve connections are replaced by numeric inputs that are typically identified as x. [Gelen sinir bağlantıları, tipik olarak x olarak tanımlanan sayısal girdilerle değiştirilir.] When there's more than one input value, x is considered a vector with elements named x1, x2, and so on. [Birden fazla girdi değeri olduğunda, x, x1, x2 vb. adlı öğelere sahip bir vektör olarak kabul edilir.]

Associated with each x value is a weight (w), which is used to strengthen or weaken the effect of the x value to simulate learning. [Her x değeriyle ilişkili, öğrenmeyi simüle etmek için x değerinin etkisini güçlendirmek veya zayıflatmak için kullanılan bir ağırlıktır (w).] Additionally, a bias (b) input is added to enable fine-grained control over the network. [Ek olarak, ağ üzerinde hassas kontrol sağlamak için bir önyargı (b) girişi eklenir.] During the training process, the w and b values will be adjusted to tune the network so that it "learns" to produce correct outputs. [Eğitim süreci sırasında, w ve b değerleri, ağın doğru çıktıları üretmeyi 'öğrenmesi' için ayarlamak üzere ayarlanacaktır.]

The neuron itself encapsulates a function that calculates a weighted sum of x, w, and b. [Nöronun kendisi, x, w ve b'nin ağırlıklı toplamını hesaplayan bir fonksiyonu içine alır.] This function is in turn enclosed in an activation function that constrains the result (often to a value between 0 and 1) to determine whether or not the neuron passes an output onto the next layer of neurons in the network. [Bu fonksiyon, nöronun ağdaki bir sonraki nöron katmanına bir çıktı geçirip geçirmediğini belirlemek için sonucu (genellikle 0 ile 1 arasında bir değerle) sınırlayan bir aktivasyon fonksiyonu içine alınır.]
