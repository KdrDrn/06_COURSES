## Evaluate classification models

The training accuracy of a classification model is much less important than how well that model will work when given new, unseen data. [Bir sınıflandırma modelinin eğitim doğruluğu, yeni, görünmeyen veriler verildiğinde bu modelin ne kadar iyi çalışacağından çok daha az önemlidir.] After all, we train models so that they can be used on new data we find in the real world. [Sonuçta, modelleri gerçek dünyada bulduğumuz yeni veriler üzerinde kullanılabilecek şekilde eğitiyoruz.] So, after we have trained a classification model, we should evaluate how it performs on a set of new, unseen data. [Bu nedenle, bir sınıflandırma modelini eğittikten sonra, bir dizi yeni, görünmeyen veri üzerinde nasıl performans gösterdiğini değerlendirmeliyiz.]

In the previous units, we created a model that would predict whether a patient had diabetes or not based on their blood glucose level. [Önceki ünitelerde, bir hastanın şeker hastası olup olmadığını kan şekeri düzeyine göre tahmin edecek bir model oluşturduk.] Now, when applied to some data that wasn't part of the training set we get the following predictions: [Şimdi, eğitim setinin parçası olmayan bazı verilere uygulandığında aşağıdaki tahminleri alıyoruz:]

Recall that x refers to blood glucose level, y refers to whether they're actually diabetic, and ŷ refers to the model’s prediction as to whether they're diabetic or not. [x'in kan şekeri seviyesini, y'nin gerçekten diyabetik olup olmadıklarını ve ŷ'nin modelin diyabetik olup olmadıklarına ilişkin tahminini ifade ettiğini hatırlayın.]

Simply calculating how many predictions were correct is sometimes misleading or too simplistic for us to understand the kinds of errors it will make in the real world. [Sadece kaç tahminin doğru olduğunu hesaplamak bazen yanıltıcı olabilir veya gerçek dünyada yapacağı hataları anlamamız için çok basittir.] To get more detailed information, we can tabulate the results in a structure called a confusion matrix, like this: [Daha detaylı bilgi almak için sonuçları aşağıdaki gibi bir karışıklık matrisi adı verilen bir yapıda tablolaştırabiliriz:]

The confusion matrix shows the total number of cases where:

- The model predicted 0 and the actual label is 0 (true negatives; top left)

- The model predicted 1 and the actual label is 1 (true positives; bottom right)

- The model predicted 0 and the actual label is 1 (false negatives; bottom left)

- The model predicted 1 and the actual label is 0 (false positives; top right)

The cells in the confusion matrix are often shaded so that higher values have a deeper shade. [Karışıklık matrisindeki hücreler genellikle gölgelenir, böylece daha yüksek değerler daha derin bir gölgeye sahip olur.] This makes it easier to see a strong diagonal trend from top-left to bottom-right, highlighting the cells where the predicted value and actual value are the same. [Bu, tahmin edilen değer ile gerçek değerin aynı olduğu hücreleri vurgulayarak sol üstten sağ alta doğru güçlü bir çapraz trend görmeyi kolaylaştırır.]

From these core values, you can calculate a range of other metrics that can help you evaluate the performance of the model. [Bu temel değerlerden, modelin performansını değerlendirmenize yardımcı olabilecek bir dizi başka metrik hesaplayabilirsiniz.] For example: [Örneğin:]

- **Accuracy**: (TP+TN)/(TP+TN+FP+FN) - out all of the predictions, how many were correct?

- **Recall**: TP/(TP+FN) - of all the cases that are positive, how many did the model identify?

- **Precision**: TP/(TP+FP) - of all the cases that the model predicted to be positive, how many actually are positive?
