## Introduction

Classification is a form of machine learning in which you train a model to predict which category an item belongs to. [Sınıflandırma, bir öğenin hangi kategoriye ait olduğunu tahmin etmek için bir modeli eğittiğiniz bir makine öğrenimi biçimidir.] For example, a health clinic might use diagnostic data such as a patient's height, weight, blood pressure, blood-glucose level to predict whether or not the patient is diabetic. [Örneğin, bir sağlık kliniği, hastanın diyabetik olup olmadığını tahmin etmek için hastanın boyu, kilosu, kan basıncı, kan şekeri seviyesi gibi tanı verilerini kullanabilir.]

Categorical data has distinct 'classes', rather than numeric values. [Kategorik verilerin sayısal değerler yerine farklı 'sınıfları' vardır.] Some kinds of data can be either numeric or categorical: the time to run a race could be a time in seconds, or we could split times into classes of ‘fast’, ‘medium’ and ‘slow’ - categorical. [Bazı veri türleri hem sayısal hem de kategorik olabilir: bir yarış için gereken süre saniye cinsinden bir süre olabilir veya süreleri 'hızlı', 'orta' ve 'yavaş' - kategorik olarak sınıflara ayırabiliriz.] While other kinds of data can only be categorical, such as a type of shape - ‘circle’, ‘triangle’, or ‘square’. [Diğer veri türleri yalnızca kategorik olabilir, örneğin bir şekil türü - 'daire', 'üçgen' veya 'kare'.]
