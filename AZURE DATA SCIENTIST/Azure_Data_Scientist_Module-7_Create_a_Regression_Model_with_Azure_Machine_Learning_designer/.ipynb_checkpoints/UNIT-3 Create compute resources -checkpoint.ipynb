## Create compute resources 

To train and deploy models using Azure Machine Learning designer, you need compute targets to run the training process. [Azure Machine Learning tasarımcısını kullanarak modelleri eğitmek ve dağıtmak için eğitim sürecini çalıştırmak üzere işlem hedeflerine ihtiyacınız vardır.] You will also use these compute targets to test the trained model after its deployment. [Bu işlem hedeflerini, dağıtıldıktan sonra eğitilen modeli test etmek için de kullanacaksınız.]

### Create compute targets

Compute targets are cloud-based resources on which you can run model training and data exploration processes. [İşlem hedefleri, üzerinde model eğitimi ve veri araştırma süreçleri çalıştırabileceğiniz bulut tabanlı kaynaklardır.]

In Azure Machine Learning studio, view the Compute page (under Manage). [Azure Machine Learning stüdyosunda, Hesaplama sayfasını görüntüleyin (Yönet altında).] This is where you manage the compute targets for your data science activities. [Burası, veri bilimi etkinlikleriniz için işlem hedeflerini yönettiğiniz yerdir.] There are four kinds of compute resource you can create: [Oluşturabileceğiniz dört tür işlem kaynağı vardır:]

- Compute Instances: Development workstations that data scientists can use to work with data and models. [Hesaplama Örnekleri: Veri bilimcilerin veri ve modellerle çalışmak için kullanabileceği geliştirme iş istasyonları.]

- Compute Clusters: Scalable clusters of virtual machines for on-demand processing of experiment code. [Hesaplama Kümeleri: Deney kodunun isteğe bağlı işlenmesi için ölçeklenebilir sanal makine kümeleri.]

- Inference Clusters: Deployment targets for predictive services that use your trained models. [Çıkarım Kümeleri: Eğitilmiş modellerinizi kullanan tahmine dayalı hizmetler için dağıtım hedefleri.]

- Attached Compute: Links to existing Azure compute resources, such as Virtual Machines or Azure Databricks clusters. [Ekli İşlem: Sanal Makineler veya Azure Databricks kümeleri gibi mevcut Azure işlem kaynaklarına bağlantılar.]
 
**Note**

Compute instances and clusters are based on standard Azure virtual machine images. [İşlem örnekleri ve kümeler, standart Azure sanal makine görüntülerini temel alır.] For this module, the Standard_DS11_v2 image is recommended to achieve the optimal balance of cost and performance. [Bu modül için, optimum maliyet ve performans dengesini elde etmek için Standard_DS11_v2 görüntüsü önerilir.] If your subscription has a quota that does not include this image, choose an alternative image; but bear in mind that a larger image may incur higher cost and a smaller image may not be sufficient to complete the tasks. [Aboneliğinizin bu görseli içermeyen bir kotası varsa, alternatif bir görsel seçin; ancak daha büyük bir görüntünün daha yüksek maliyete neden olabileceğini ve görevleri tamamlamak için daha küçük bir görüntünün yeterli olmayabileceğini unutmayın.] Alternatively, ask your Azure administrator to extend your quota. [Alternatif olarak, Azure yöneticinizden kotanızı genişletmesini isteyin.]

1. On the Compute Instances tab, add a new compute instance with the following settings. [İşlem Örnekleri sekmesinde, aşağıdaki ayarlarla yeni bir işlem örneği ekleyin.] You'll use this as a workstation from which to test your model: [Bunu, modelinizi test etmek için bir iş istasyonu olarak kullanacaksınız:]

- Compute name: enter a unique name [Hesaplama adı: benzersiz bir ad girin]

- Virtual Machine type: CPU [Sanal Makine türü: CPU]

- Virtual Machine size: [Sanal Makine boyutu:]

    - Choose Select from all options [Tüm seçeneklerden Seç'i seçin]

    - Search for and select Standard_DS11_v2 [Standard_DS11_v2'yi arayın ve seçin]

2. While the compute instance is being created, switch to the Compute Clusters tab, and add a new compute cluster with the following settings. [İşlem örneği oluşturulurken İşlem Kümeleri sekmesine geçin ve aşağıdaki ayarlarla yeni bir işlem kümesi ekleyin.] You'll use this to train a machine learning model: [Bunu bir makine öğrenimi modelini eğitmek için kullanacaksınız:]

- Location: Select the same as your workspace. [Konum: Çalışma alanınızla aynısını seçin.] If that location is not listed, choose the one closest to you [Bu konum listelenmemişse, size en yakın olanı seçin]

- Virtual Machine tier: Dedicated [Sanal Makine katmanı: Özel]

- Virtual Machine type: CPU [Sanal Makine türü: CPU]

- Virtual Machine size: [Sanal Makine boyutu:]

    - Choose Select from all options [Tüm seçeneklerden Seç'i seçin]

    - Search for and select Standard_DS11_v2 [Standard_DS11_v2'yi arayın ve seçin]

- Compute name: enter a unique name [Hesaplama adı: benzersiz bir ad girin]

- Minimum number of nodes: 0 [Minimum düğüm sayısı: 0]

- Maximum number of nodes: 2 [Maksimum düğüm sayısı: 2]

- Idle seconds before scale down: 120 [Ölçeğin küçültülmesinden önceki boşta saniye: 120]

- Enable SSH access: Unselected [SSH erişimini etkinleştir: Seçili değil]
 
**Tip**

After completing each module, be sure to follow the Clean Up instructions at the end of the module to stop your compute resources. [Her modülü tamamladıktan sonra, bilgi işlem kaynaklarınızı durdurmak için modülün sonundaki Temizleme yönergelerini izlediğinizden emin olun.] Stopping your compute ensures your subscription won't be charged for compute resources. [İşleminizi durdurmak, aboneliğinizin işlem kaynakları için ücretlendirilmemesini sağlar.]

The compute targets will take some time to be created. [İşlem hedeflerinin oluşturulması biraz zaman alacaktır.] You can move onto the next unit while you wait. [Beklerken bir sonraki üniteye geçebilirsiniz.]
