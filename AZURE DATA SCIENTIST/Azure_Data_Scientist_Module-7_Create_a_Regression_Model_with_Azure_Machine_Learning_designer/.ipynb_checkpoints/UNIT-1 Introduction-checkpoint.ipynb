### Introduction

Regression is a form of machine learning that is used to predict a numeric label based on an item's features. [Regresyon, bir öğenin özelliklerine dayalı olarak sayısal bir etiketi tahmin etmek için kullanılan bir makine öğrenimi biçimidir.] For example, an automobile sales company might use the characteristics of a car (such as engine size, number of seats, mileage, and so on) to predict its likely selling price. [Örneğin, bir otomobil satış şirketi, olası satış fiyatını tahmin etmek için bir arabanın özelliklerini (motor boyutu, koltuk sayısı, kilometre vb.) kullanabilir.] In this case, the characteristics of the car are the features, and the selling price is the label. [Bu durumda, arabanın özellikleri özelliklerdir ve satış fiyatı etikettir.]



Regression is an example of a supervised machine learning technique in which you train a model using data that includes both the features and known values for the label, so that the model learns to fit the feature combinations to the label. [Regresyon, etiket için hem özellikleri hem de bilinen değerleri içeren verileri kullanarak bir modeli eğittiğiniz, böylece modelin özellik kombinasyonlarını etikete uydurmayı öğrendiği, denetimli bir makine öğrenimi tekniğinin bir örneğidir.] Then, after training has been completed, you can use the trained model to predict labels for new items for which the label is unknown. [Ardından, eğitim tamamlandıktan sonra, etiketi bilinmeyen yeni öğeler için etiketleri tahmin etmek için eğitilmiş modeli kullanabilirsiniz.]

You can use Microsoft Azure Machine Learning designer to create regression models by using a drag and drop visual interface, without needing to write any code. [Microsoft Azure Machine Learning tasarımcısını herhangi bir kod yazmaya gerek kalmadan sürükle ve bırak görsel arabirimi kullanarak regresyon modelleri oluşturmak için kullanabilirsiniz.]

In this module, you'll learn how to: [Bu modülde şunları nasıl yapacağınızı öğreneceksiniz:]

- Use Azure Machine Learning designer to train a regression model. [Bir regresyon modelini eğitmek için Azure Machine Learning tasarımcısını kullanın.]

- Use a regression model for inferencing. [Çıkarım yapmak için bir regresyon modeli kullanın.]

Deploy a regression model as a service. [Hizmet olarak bir regresyon modeli dağıtın.]
