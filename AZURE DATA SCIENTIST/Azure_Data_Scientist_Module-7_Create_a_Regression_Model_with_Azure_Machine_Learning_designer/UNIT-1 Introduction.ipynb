### Introduction

Regression is a form of machine learning that is used to predict a numeric label based on an item's features. [Regresyon, bir enin zelliklerine dayal olarak saysal bir etiketi tahmin etmek iin kullanlan bir makine renimi biimidir.] For example, an automobile sales company might use the characteristics of a car (such as engine size, number of seats, mileage, and so on) to predict its likely selling price. [rnein, bir otomobil sat irketi, olas sat fiyatn tahmin etmek iin bir arabann zelliklerini (motor boyutu, koltuk says, kilometre vb.) kullanabilir.] In this case, the characteristics of the car are the features, and the selling price is the label. [Bu durumda, arabann zellikleri zelliklerdir ve sat fiyat etikettir.]



Regression is an example of a supervised machine learning technique in which you train a model using data that includes both the features and known values for the label, so that the model learns to fit the feature combinations to the label. [Regresyon, etiket iin hem zellikleri hem de bilinen deerleri ieren verileri kullanarak bir modeli eittiiniz, bylece modelin zellik kombinasyonlarn etikete uydurmay rendii, denetimli bir makine renimi tekniinin bir rneidir.] Then, after training has been completed, you can use the trained model to predict labels for new items for which the label is unknown. [Ardndan, eitim tamamlandktan sonra, etiketi bilinmeyen yeni eler iin etiketleri tahmin etmek iin eitilmi modeli kullanabilirsiniz.]
