## Deploy a predictive service

In Azure Machine Learning, you can deploy a service as an Azure Container Instances (ACI) or to an Azure Kubernetes Service (AKS) cluster. [Azure Machine Learning, bir hizmeti Azure Container Instances (ACI) olarak veya bir Azure Kubernetes Service (AKS) kümesine dağıtabilirsiniz.] For production scenarios, an AKS deployment is recommended, for which you must create an inference cluster compute target. [Üretim senaryoları için, bir çıkarım kümesi işlem hedefi oluşturmanız gereken bir AKS dağıtımı önerilir.] In this exercise, you'll use an ACI service, which is a suitable deployment target for testing, and does not require you to create an inference cluster. [Bu alıştırmada, test için uygun bir dağıtım hedefi olan ve bir çıkarım kümesi oluşturmanızı gerektirmeyen bir ACI hizmeti kullanacaksınız.]

1. In Azure Machine Learning studio, on the **Automated ML** page, select the run for your automated machine learning experiment. [Azure Machine Learning stüdyosunda, Otomatik ML sayfasında, otomatikleştirilmiş makine öğrenimi denemeniz için çalıştırmayı seçin.]

2. On the **Details** tab, select the algorithm name for the best model. [Ayrıntılar sekmesinde, en iyi model için algoritma adını seçin.]



3. On the **Model** tab, select the **Deploy** button and use the **Deploy to web service** option to deploy the model with the following settings: [Model sekmesinde, Dağıt düğmesini seçin ve modeli aşağıdaki ayarlarla dağıtmak için Web hizmetine dağıt seçeneğini kullanın:]

    - **Name**: predict-rentals [İsim: tahmin kiralama]

    - **Description**: Predict cycle rentals [Açıklama: Bisiklet kiralamalarını tahmin edin]

    - **Compute type**: Azure Container Instance [İşlem türü: Azure Container Instance]

    - **Enable authentication**: Selected [Kimlik doğrulamayı etkinleştir: Seçildi]

4. Wait for the deployment to start - this may take a few seconds. [Dağıtımın başlamasını bekleyin - bu işlem birkaç saniye sürebilir.] Then, in the **Model summary** section, observe the **Deploy status** for the **predict-rentals service**, which should be **Running**. [Ardından, Model özeti bölümünde, Çalışıyor olması gereken tahmin kiralama hizmeti için Dağıtım durumunu gözlemleyin.] Wait for this status to change to **Successful**. [Bu durumun Başarılı olarak değişmesini bekleyin.] You may need to select **↻ Refresh** periodically. [↻ Periyodik olarak yenile'yi seçmeniz gerekebilir.]

5. In Azure Machine Learning studio, view the **Endpoints** page and select the **predict-rentals** real-time endpoint. [Azure Machine Learning stüdyosunda, Uç Noktalar sayfasını görüntüleyin ve tahmin kiralama gerçek zamanlı uç noktasını seçin.] Then select the **Consume** tab and note the following information there. [Ardından Tüket sekmesini seçin ve oradaki aşağıdaki bilgileri not edin.] If you do not see the **Consume** tab, the deployment is not completely finished, you will need to wait and refresh the page. [Tüket sekmesini görmüyorsanız, dağıtım tamamen bitmemiştir, beklemeniz ve sayfayı yenilemeniz gerekecektir.] You need the information from the **Consume** tab to connect to your deployed service from a client application. [Dağıtılan hizmetinize bir istemci uygulamasından bağlanmak için Tüket sekmesindeki bilgilere ihtiyacınız vardır.]

    - The REST endpoint for your service [Hizmetiniz için REST uç noktası]

    - The Primary Key for your service [hizmetiniz için Birincil Anahtar]
    


6. Note that you can use the ⧉ link next to these values to copy them to the clipboard. [Bunları panoya kopyalamak için bu değerlerin yanındaki ⧉ bağlantısını kullanabileceğinizi unutmayın.]
