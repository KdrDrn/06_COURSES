## Deploy a predictive service

In Azure Machine Learning, you can deploy a service as an Azure Container Instances (ACI) or to an Azure Kubernetes Service (AKS) cluster. [Azure Machine Learning, bir hizmeti Azure Container Instances (ACI) olarak veya bir Azure Kubernetes Service (AKS) kmesine databilirsiniz.] For production scenarios, an AKS deployment is recommended, for which you must create an inference cluster compute target. [retim senaryolar iin, bir karm kmesi ilem hedefi oluturmanz gereken bir AKS datm nerilir.] In this exercise, you'll use an ACI service, which is a suitable deployment target for testing, and does not require you to create an inference cluster. [Bu altrmada, test iin uygun bir datm hedefi olan ve bir karm kmesi oluturmanz gerektirmeyen bir ACI hizmeti kullanacaksnz.]

1. In Azure Machine Learning studio, on the **Automated ML** page, select the run for your automated machine learning experiment. [Azure Machine Learning stdyosunda, Otomatik ML sayfasnda, otomatikletirilmi makine renimi denemeniz iin altrmay sein.]

2. On the **Details** tab, select the algorithm name for the best model. [Ayrntlar sekmesinde, en iyi model iin algoritma adn sein.]



3. On the **Model** tab, select the **Deploy** button and use the **Deploy to web service** option to deploy the model with the following settings: [Model sekmesinde, Dat dmesini sein ve modeli aadaki ayarlarla datmak iin Web hizmetine dat seeneini kullann:]

    - **Name**: predict-rentals [sim: tahmin kiralama]

    - **Description**: Predict cycle rentals [Aklama: Bisiklet kiralamalarn tahmin edin]

    - **Compute type**: Azure Container Instance [lem tr: Azure Container Instance]

    - **Enable authentication**: Selected [Kimlik dorulamay etkinletir: Seildi]

4. Wait for the deployment to start - this may take a few seconds. [Datmn balamasn bekleyin - bu ilem birka saniye srebilir.] Then, in the **Model summary** section, observe the **Deploy status** for the **predict-rentals service**, which should be **Running**. [Ardndan, Model zeti blmnde, alyor olmas gereken tahmin kiralama hizmeti iin Datm durumunu gzlemleyin.] Wait for this status to change to **Successful**. [Bu durumun Baarl olarak deimesini bekleyin.] You may need to select ** Refresh** periodically. [ Periyodik olarak yenile'yi semeniz gerekebilir.]

5. In Azure Machine Learning studio, view the **Endpoints** page and select the **predict-rentals** real-time endpoint. [Azure Machine Learning stdyosunda, U Noktalar sayfasn grntleyin ve tahmin kiralama gerek zamanl u noktasn sein.] Then select the **Consume** tab and note the following information there. [Ardndan Tket sekmesini sein ve oradaki aadaki bilgileri not edin.] If you do not see the **Consume** tab, the deployment is not completely finished, you will need to wait and refresh the page. [Tket sekmesini grmyorsanz, datm tamamen bitmemitir, beklemeniz ve sayfay yenilemeniz gerekecektir.] You need the information from the **Consume** tab to connect to your deployed service from a client application. [Datlan hizmetinize bir istemci uygulamasndan balanmak iin Tket sekmesindeki bilgilere ihtiyacnz vardr.]

    - The REST endpoint for your service [Hizmetiniz iin REST u noktas]
