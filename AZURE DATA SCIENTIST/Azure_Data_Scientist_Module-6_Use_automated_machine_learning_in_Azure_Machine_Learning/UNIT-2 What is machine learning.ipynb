## What is machine learning?

Machine learning is a technique that uses mathematics and statistics to create a model that can predict unknown values. [Makine öğrenimi, bilinmeyen değerleri tahmin edebilen bir model oluşturmak için matematik ve istatistik kullanan bir tekniktir.]

For example, suppose Adventure Works Cycles is a business that rents cycles in a city. [Örneğin, Adventure Works Cycles'ın bir şehirde bisiklet kiralayan bir işletme olduğunu varsayalım.] The business could use historic data to train a model that predicts daily rental demand in order to make sure sufficient staff and cycles are available. [İşletme, yeterli personel ve döngünün mevcut olduğundan emin olmak için günlük kiralama talebini tahmin eden bir modeli eğitmek için geçmiş verileri kullanabilir.]

To do this, Adventure Works could create a machine learning model that takes information about a specific day (the day of week, the anticipated weather conditions, and so on) as an input, and predicts the expected number of rentals as an output. [Bunu yapmak için Adventure Works, girdi olarak belirli bir gün (haftanın günü, beklenen hava koşulları vb.) hakkında bilgi alan ve çıktı olarak beklenen kiralama sayısını tahmin eden bir makine öğrenimi modeli oluşturabilir.]

Mathematically, you can think of machine learning as a way of defining a function (let's call it f) that operates on one or more features of something (which we'll call x) to calculate a predicted label (y) - like this: [Matematiksel olarak, makine öğrenimini, tahmin edilen bir etiketi (y) hesaplamak için bir şeyin (x olarak adlandıracağız) bir veya daha fazla özelliği üzerinde çalışan bir işlevi (haydi f diyelim) tanımlamanın bir yolu olarak düşünebilirsiniz - şöyle:]

f(x) = y

In this bicycle rental example, the details about a given day (day of the week, weather, and so on) are the features (x), the number of rentals for that day is the label (y), and the function (f) that calculates the number of rentals based on the information about the day is encapsulated in a machine learning model. [Bu bisiklet kiralama örneğinde, belirli bir güne ilişkin ayrıntılar (haftanın günü, hava durumu vb.) özellikler (x), o gün için kiralama sayısı etiket (y) ve işlevdir (f). ) güne ait bilgilere göre kiralama sayısını hesaplayan bir makine öğrenmesi modelinde kapsüllenir.]

The specific operation that the f function performs on x to calculate y depends on a number of factors, including the type of model you're trying to create and the specific algorithm used to train the model. [f işlevinin y'yi hesaplamak için x üzerinde gerçekleştirdiği özel işlem, oluşturmaya çalıştığınız modelin türü ve modeli eğitmek için kullanılan özel algoritma dahil olmak üzere bir dizi faktöre bağlıdır.] Additionally in most cases, the data used to train the machine learning model requires some pre-processing before model training can be performed. [Ek olarak, çoğu durumda, makine öğrenimi modelini eğitmek için kullanılan veriler, model eğitimi gerçekleştirilmeden önce bazı ön işlemler gerektirir.]

### Azure Machine Learning
Training and deploying an effective machine learning model involves a lot of work, much of it time-consuming and resource-intensive. [Etkili bir makine öğrenimi modelini eğitmek ve dağıtmak, çoğu zaman alan ve yoğun kaynak gerektiren çok fazla iş gerektirir.] Azure Machine Learning is a cloud-based service that helps simplify some of the tasks and reduce the time it takes to prepare data, train a model, and deploy a predictive service. [Azure Machine Learning, bazı görevleri basitleştirmeye yardımcı olan ve veri hazırlama, model eğitme ve tahmine dayalı bir hizmet dağıtma süresini kısaltmaya yardımcı olan bulut tabanlı bir hizmettir.] In the rest of this unit, you'll explore Azure Machine Learning, and in particular its automated machine learning capability. [Bu ünitenin geri kalanında Azure Machine Learning'i ve özellikle otomatik makine öğrenimi özelliğini keşfedeceksiniz.]
