## Introduction

Classification is a form of machine learning that is used to predict which category, or class, an item belongs to. [Sınıflandırma, bir öğenin hangi kategoriye veya sınıfa ait olduğunu tahmin etmek için kullanılan bir makine öğrenimi biçimidir.] For example, a health clinic might use the characteristics of a patient (such as age, weight, blood pressure, and so on) to predict whether the patient is at risk of diabetes. [Örneğin bir sağlık kliniği, hastanın diyabet riski altında olup olmadığını tahmin etmek için hastanın özelliklerini (yaş, kilo, tansiyon vb.) kullanabilir.] In this case, the characteristics of the patient are the features, and the label is a classification of either 0 or 1, representing non-diabetic or diabetic. [Bu durumda, hastanın özellikleri özelliklerdir ve etiket, diyabetik olmayanları veya diyabetikleri temsil eden 0 veya 1 sınıflandırmasıdır.]



Classification is an example of a supervised machine learning technique in which you train a model using data that includes both the features and known values for the label, so that the model learns to fit the feature combinations to the label. [Sınıflandırma, etiket için hem özellikleri hem de bilinen değerleri içeren verileri kullanarak bir modeli eğittiğiniz, böylece modelin özellik kombinasyonlarını etikete uydurmayı öğrendiği, denetimli bir makine öğrenimi tekniğinin bir örneğidir.] Then, after training has been completed, you can use the trained model to predict labels for new items for which the label is unknown. [Ardından, eğitim tamamlandıktan sonra, etiketi bilinmeyen yeni öğeler için etiketleri tahmin etmek için eğitilmiş modeli kullanabilirsiniz.]

You can use Microsoft Azure Machine Learning designer to create classification models by using a drag and drop visual interface, without needing to write any code. [Herhangi bir kod yazmaya gerek kalmadan, sürükle ve bırak görsel arabirimi kullanarak sınıflandırma modelleri oluşturmak için Microsoft Azure Machine Learning tasarımcısını kullanabilirsiniz.]

In this module, you'll learn how to: [Bu modülde şunları nasıl yapacağınızı öğreneceksiniz:]

- Use Azure Machine Learning designer to train a classification model. [Bir sınıflandırma modelini eğitmek için Azure Machine Learning tasarımcısını kullanın.]

- Use a classification model for inferencing. [Çıkarım yapmak için bir sınıflandırma modeli kullanın.]

- Deploy a classification model as a service. [Hizmet olarak bir sınıflandırma modeli dağıtın.]

To complete this module, you'll need a Microsoft Azure subscription. [Henüz sahip değilseniz, https://azure.microsoft.com adresinde ücretsiz deneme için kaydolabilirsiniz.]
