### Welcome to the AWS Machine Learning Foundations Course!

**Course Overview** 

- **Lesson 2: Introduction to Machine Learning** – In this lesson, you will learn the fundamentals of supervised and unsupervised machine learning, including the process steps of solving machine learning problems, and explore several examples. [Ders 2: Makine Öğrenimine Giriş – Bu derste, makine öğrenimi sorunlarını çözme işlem adımları da dahil olmak üzere, denetimli ve denetimsiz makine öğreniminin temellerini öğrenecek ve çeşitli örnekleri keşfedeceksiniz.]
- **Lesson 3: Machine Learning with AWS** – In this lesson, you will learn about advanced machine learning techniques such as generative AI and reinforcement learning. [Ders 3: AWS ile Makine Öğrenimi – Bu derste, üretken yapay zeka ve pekiştirmeli öğrenme gibi gelişmiş makine öğrenimi teknikleri hakkında bilgi edineceksiniz.] You will also learn how to train these models with AWS AI/ML services. [Ayrıca bu modelleri AWS AI/ML hizmetleriyle nasıl eğiteceğinizi öğreneceksiniz.]
- **Lesson 4: Software Engineering Practices, part 1** – In this lesson, you will learn how to write well-documented, modularized code. [Ders 4: Yazılım Mühendisliği Uygulamaları, bölüm 1 – Bu derste, iyi belgelenmiş, modülerleştirilmiş kodun nasıl yazılacağını öğreneceksiniz.]
- **Lesson 5: Software Engineering Practices, part 2** – In this lesson, you will learn how to test your code and log best practices. [Ders 5: Yazılım Mühendisliği Uygulamaları, bölüm 2 – Bu derste, kodunuzu nasıl test edeceğinizi ve en iyi uygulamaları nasıl günlüğe kaydedeceğinizi öğreneceksiniz.]
- **Lesson 6: Object-Oriented Programming** – In this lesson, you will learn about this programming style and prepare to write your own Python package. [Ders 6: Nesneye Yönelik Programlama – Bu derste, bu programlama stilini öğrenecek ve kendi Python paketinizi yazmaya hazırlanacaksınız.]

**By the end of the course, you will be able to...**

- Explain machine learning and the types of questions machine learning can help to solve. [Makine öğrenimini ve makine öğreniminin çözülmesine yardımcı olabileceği soru türlerini açıklayın.]
- Explain what machine learning solutions AWS offers and how AWS AI devices put machine learning in the hands of every developer. [AWS'nin hangi makine öğrenimi çözümlerini sunduğunu ve AWS AI cihazlarının makine öğrenimini her geliştiricinin eline nasıl verdiğini açıklayın.]
- Apply software engineering principles of modular code, code efficiency, refactoring, documentation, and version control to data science. [Modüler kod, kod verimliliği, yeniden düzenleme, dokümantasyon ve sürüm kontrolünün yazılım mühendisliği ilkelerini veri bilimine uygulayın.]
- Apply software engineering principles of testing code, logging, and conducting code reviews to data science. [Kod test etme, günlüğe kaydetme ve kod incelemeleri yürütmeye ilişkin yazılım mühendisliği ilkelerini veri bilimine uygulayın.]
- Implement the basic principles of object-oriented programming to build a Python package. [Bir Python paketi oluşturmak için nesne yönelimli programlamanın temel ilkelerini uygulayın.]
