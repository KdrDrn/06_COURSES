<font color="lightseagreen" size=+2.5><b>4. Reinforcement Learning and Its Applications</b></font>

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" style="color:white" data-toggle="popover">Table of Contents</a>

This section introduces you to a type of machine learning (ML) called reinforcement learning (RL). [Bu bölüm size pekiştirmeli öğrenme (RL) adı verilen bir tür makine öğrenimi (ML) tanıtır.] You'll hear about its real-world applications and learn basic concepts using AWS DeepRacer as an example. [Gerçek dünyadaki uygulamaları duyacak ve örnek olarak AWS DeepRacer kullanarak temel kavramları öğreneceksiniz.] By the end of the section, you will be able to create, train, and evaluate a reinforcement learning model in the AWS DeepRacer console. [Bölümün sonunda, AWS DeepRacer konsolunda bir pekiştirmeli öğrenme modeli oluşturabilecek, eğitebilecek ve değerlendirebileceksiniz.]

**Introduction to Reinforcement Learning**











**Summary**

In reinforcement learning (RL), an agent is trained to achieve a goal based on the feedback it receives as it interacts with an environment. [Takviyeli öğrenmede (RL), bir aracı, bir ortamla etkileşime girerken aldığı geri bildirime dayalı olarak bir hedefe ulaşmak için eğitilir.] It collects a number as a reward for each action it takes. [Yaptığı her eylem için ödül olarak bir sayı toplar.] Actions that help the agent achieve its goal are incentivized with higher numbers. [Temsilcinin amacına ulaşmasına yardımcı olan eylemler daha yüksek sayılarla teşvik edilir.] Unhelpful actions result in a low reward or no reward. [Yararsız eylemler, düşük bir ödülle veya hiç ödülle sonuçlanmaz.]

With a learning objective of **maximizing total cumulative reward**, over time, the agent learns, through trial and error, to map gainful actions to situations. [Zaman içinde toplam kümülatif ödülü maksimize etmeye yönelik bir öğrenme hedefiyle aracı, deneme yanılma yoluyla kazançlı eylemleri durumlarla eşleştirmeyi öğrenir.] The better trained the agent, the more efficiently it chooses actions that accomplish its goal. [Ajan ne kadar iyi eğitilirse, amacına ulaşan eylemleri o kadar verimli seçer.]

**Reinforcement Learning Applications**

**Summary**

Reinforcement learning is used in a variety of fields to solve real-world problems. It’s particularly useful for addressing sequential problems with long-term goals. Let’s take a look at some examples.

1. RL is great at **playing games**:
- **Go** (board game) was mastered by the AlphaGo Zero software.
- **Atari classic video** games are commonly used as a learning tool for creating and testing RL software.
- **StarCraft II**, the real-time strategy video game, was mastered by the AlphaStar software.

2. RL is used in **video game level design**:
- Video game level design determines how complex each stage of a game is and directly affects how boring, frustrating, or fun it is to play that game.
- Video game companies create an agent that plays the game over and over again to collect data that can be visualized on graphs.
- This visual data gives designers a quick way to assess how easy or difficult it is for a player to make progress, which enables them to find that “just right” balance between boredom and frustration faster.
3. RL is used in **wind energy optimization**:
- RL models can also be used to power robotics in physical devices.
- When multiple turbines work together in a wind farm, the turbines in the front, which receive the wind first, can cause poor wind conditions for the turbines behind them. This is called **wake turbulence** and it reduces the amount of energy that is captured and converted into electrical power.
- Wind energy organizations around the world use reinforcement learning to test solutions. Their models respond to changing wind conditions by changing the angle of the turbine blades. When the upstream turbines slow down it helps the downstream turbines capture more energy.

4. Other examples of real-world RL include:
- **Industrial robotics**
- **Fraud detection**
- **Stock trading**
- **Autonomous driving**



**New Terms**
- **Agent**: The piece of software you are training is called an agent. It makes decisions in an environment to reach a goal.
- **Environment**: The environment is the surrounding area with which the agent interacts.
- **Reward**: Feedback is given to an agent for each action it takes in a given state. This feedback is a numerical reward.
- **Action**: For every state, an agent needs to take an action toward achieving its goal.

## <a id="5"></a>
<font color="lightseagreen" size=+2.5><b>5. Reinforcement Learning with AWS DeepRacer</b></font>

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" style="color:white" data-toggle="popover">Table of Contents</a>

**Reinforcement Learning Concepts**

In this section, we’ll learn some basic reinforcement learning terms and concepts using AWS DeepRacer as an example. [Bu bölümde, örnek olarak AWS DeepRacer kullanarak bazı temel pekiştirmeli öğrenme terimlerini ve kavramlarını öğreneceğiz.]

**Summary**

This section introduces six basic reinforcement learning terms and provides an example for each in the context of AWS DeepRacer. [Bu bölümde altı temel pekiştirmeli öğrenme terimi tanıtılır ve her biri için AWS DeepRacer bağlamında bir örnek sağlanır.]



**Agent**
- The piece of software you are training is called an agent. [Eğittiğiniz yazılım parçasına aracı denir.]
- It makes decisions in an environment to reach a goal. [Bir amaca ulaşmak için bir ortamda kararlar alır.]
- In AWS DeepRacer, the agent is the AWS DeepRacer car and its goal is to finish * laps around the track as fast as it can while, in some cases, avoiding obstacles. [AWS DeepRacer'da aracı, AWS DeepRacer arabasıdır ve amacı, bazı durumlarda engellerden kaçınırken, pistteki * turları olabildiğince hızlı bitirmektir.]

**Environment**
- The environment is the surrounding area within which our agent interacts. [Çevre, aracımızın etkileşimde bulunduğu çevre alandır.]
- For AWS DeepRacer, this is a track in our simulator or in real life. [AWS DeepRacer için bu, simülatörümüzdeki veya gerçek hayattaki bir parkurdur.]

**State**
- The state is defined by the current position within the environment that is visible, or known, to an agent. [Durum, bir ajan tarafından görülebilen veya bilinen ortam içindeki mevcut konumla tanımlanır.]
- In AWS DeepRacer’s case, each state is an image captured by its camera. [AWS DeepRacer örneğinde, her durum kendi kamerası tarafından yakalanan bir görüntüdür.]
- The car’s initial state is the starting line of the track and its terminal state is when the car finishes a lap, bumps into an obstacle, or drives off the track. [Otomobilin ilk durumu, pistin başlangıç ​​çizgisidir ve son durumu, otomobilin bir turu tamamlaması, bir engele çarpması veya pistten çıkmasıdır.]

**Action**
- For every state, an agent needs to take an action toward achieving its goal. [Her devlet için, bir ajanın amacına ulaşmak için bir eylemde bulunması gerekir.]
An AWS DeepRacer car approaching a turn can choose to accelerate or brake and turn left, right, or go straight. [Bir dönüşe yaklaşan bir AWS DeepRacer arabası, hızlanmayı veya fren yapmayı ve sola, sağa dönmeyi veya düz gitmeyi seçebilir.]

**Reward**
- Feedback is given to an agent for each action it takes in a given state. [Belirli bir durumda yaptığı her eylem için bir aracıya geri bildirim verilir.]
- This feedback is a numerical reward. [Bu geri bildirim sayısal bir ödüldür.]
- A reward function is an incentive plan that assigns scores as rewards to different zones on the track. [Ödül işlevi, puanları pistteki farklı bölgelere ödül olarak atayan bir teşvik planıdır.]

**Episode**
- An episode represents a period of trial and error when an agent makes decisions and gets feedback from its environment. [Bir bölüm, bir aracının kararlar aldığı ve çevresinden geri bildirim aldığı bir deneme yanılma dönemini temsil eder.]
- For AWS DeepRacer, an episode begins at the initial state, when the car leaves the starting position, and ends at the terminal state, when it finishes a lap, bumps into an obstacle, or drives off the track. [AWS DeepRacer için bir bölüm, araç başlangıç ​​konumundan ayrıldığında başlangıç ​​durumunda başlar ve bir turu bitirdiğinde, bir engelle karşılaştığında veya pist dışına çıktığında son durumda sona erer.]

In a reinforcement learning model, an **agent** learns in an interactive real-time **environment** by trial and error using feedback from its own **actions**. [Bir pekiştirmeli öğrenme modelinde, bir etmen, kendi eylemlerinden gelen geri bildirimi kullanarak deneme yanılma yoluyla etkileşimli bir gerçek zamanlı ortamda öğrenir.] Feedback is given in the form of **rewards**. [Geri bildirim ödül şeklinde verilir.]



**Putting Your Spin on AWS DeepRacer: The Practitioner's Role in RL**

**Summary**

AWS DeepRacer may be autonomous, but you still have an important role to play in the success of your model. [AWS DeepRacer özerk olabilir, ancak yine de modelinizin başarısında oynayacağınız önemli bir rolünüz var.] In this section, we introduce the **training algorithm, action space, hyperparameters, and reward function** and discuss how your ideas make a difference. [Bu bölümde eğitim algoritmasını, eylem alanını, hiperparametreleri ve ödül işlevini tanıtıyor ve fikirlerinizin nasıl fark yarattığını tartışıyoruz.]

1. An algorithm is a set of instructions that tells a computer what to do. [Algoritma, bir bilgisayara ne yapacağını söyleyen bir dizi talimattır.] ML is special because it enables computers to learn without being explicitly programmed to do so. [ML özeldir çünkü bilgisayarların açıkça programlanmadan öğrenmesini sağlar.]

2. The training algorithm defines your model’s learning objective, which is to maximize total cumulative reward. [Eğitim algoritması, modelinizin toplam kümülatif ödülü en üst düzeye çıkarmak olan öğrenme hedefini tanımlar.] Different algorithms have different strategies for going about this. [Farklı algoritmaların bu konuda farklı stratejileri vardır.]
- A soft actor critic (SAC) embraces exploration and is data-efficient, but can lack stability. [Yumuşak aktör eleştirmeni (SAC) keşfi benimser ve veri açısından verimlidir, ancak istikrardan yoksun olabilir.]
- A proximal policy optimization (PPO) is stable but data-hungry. [Bir yakın politika optimizasyonu (PPO) kararlıdır ancak verilere aç.]

3. An action space is the set of all valid actions, or choices, available to an agent as it interacts with an environment. [Eylem alanı, bir aracı bir ortamla etkileşime girerken kullanabileceği tüm geçerli eylemler veya seçenekler kümesidir.]
- Discrete action space represents all of an agent's possible actions for each state in a finite set of steering angle and throttle value combinations. [Ayrık eylem alanı, sonlu bir direksiyon açısı ve gaz kelebeği değeri kombinasyonları kümesindeki her durum için bir aracının tüm olası eylemlerini temsil eder.]

- Continuous action space allows the agent to select an action from a range of values that you define for each sta te. [Sürekli eylem alanı, aracının her durum için tanımladığınız bir dizi değerden bir eylem seçmesine olanak tanır.]
4. Hyperparameters are variables that control the performance of your agent during training. [Hiperparametreler, eğitim sırasında aracınızın performansını kontrol eden değişkenlerdir.] There is a variety of different categories with which to experiment. [Denemek için çeşitli farklı kategoriler vardır.] Change the values to increase or decrease the influence of different parts of your model. [Modelinizin farklı bölümlerinin etkisini artırmak veya azaltmak için değerleri değiştirin.]
- For example, the learning rate is a hyperparameter that controls how many new experiences are counted in learning at each step. [Örneğin, öğrenme oranı, her adımda öğrenmede kaç yeni deneyimin sayıldığını kontrol eden bir hiper parametredir.] A higher learning rate results in faster training but may reduce the model’s quality. [Daha yüksek bir öğrenme oranı, daha hızlı eğitim ile sonuçlanır ancak modelin kalitesini düşürebilir.]

5. The reward function's purpose is to encourage the agent to reach its goal. [Ödül işlevinin amacı, aracıyı amacına ulaşması için teşvik etmektir.] Figuring out how to reward which actions is one of your most important jobs. [Hangi eylemlerin nasıl ödüllendirileceğini bulmak, en önemli işlerinizden biridir.]

**Discriminator**: A neural network trained to differentiate between real and synthetic data. [Discriminator: A neural network trained to differentiate between real and synthetic data.]

**Discriminator loss**: Evaluates how well the discriminator differentiates between real and fake data. [Discriminator loss: Evaluates how well the discriminator differentiates between real and fake data.]

**Edit event**: When a note is either added or removed from your input track during inference. [Edit event: When a note is either added or removed from your input track during inference.]

[Exploration versus exploitation: An agent should exploit known information from previous experiences to achieve higher cumulative rewards, but it also needs to explore to gain new experiences that can be used in choosing the best actions in the future.]

**Generator**: A neural network that learns to create new data resembling the source data on which it was trained. [Generator: A neural network that learns to create new data resembling the source data on which it was trained.]

**Generator loss**: Measures how far the output data deviates from the real data present in the training dataset. [Generator loss: Measures how far the output data deviates from the real data present in the training dataset.]

**Hidden layer**: A layer that occurs between the output and input layers. [Hidden layer: A layer that occurs between the output and input layers.] Hidden layers are tailored to a specific task. [Hidden layers are tailored to a specific task.]

**Input layer**: The first layer in a neural network. [Input layer: The first layer in a neural network.] This layer receives all data that passes through the neural network. [This layer receives all data that passes through the neural network.]

**Output layer**: The last layer in a neural network.
