```python
#icecek boyutlari arasinda en fazla calori farki olan icecekleri karsilastirin
```

```python
max_calorie = df.loc[df['Calories'].idxmax()]
min_calorie = df.loc[df['Calories'].idxmin()]
pd.concat([max_calorie, min_calorie], axis=1)
```

```python
calorie_by_category_max = df.groupby('Beverage_category')['Calories'].max().sort_index().to_frame()
calorie_by_category_max.rename(columns={'Calories':'Max_Calorie'}, inplace=True)
calorie_by_category_max
```

```python
calorie_by_category_min = df.groupby('Beverage_category')['Calories'].min().sort_index().to_frame()
calorie_by_category_min.rename(columns={'Calories':'Min_Calorie'}, inplace=True)
calorie_by_category_min
```

```python
calorie_by_category = pd.merge(calorie_by_category_max, calorie_by_category_min, left_index=True, right_index=True)
calorie_by_category
```

```python
calorie_by_category['Diff'] = calorie_by_category['Max_Calorie'] - calorie_by_category['Min_Calorie']
calorie_by_category
```

```python
calorie_by_category.sort_values(by='Diff', ascending=False)
```

```python
morning = df[df['Beverage']=='Brewed Coffee']['Calories'].mean()
noon = df[df['Beverage']=='Hot Chocolate (Without Whipped Cream)']['Calories'].mean()
evening = df[df['Beverage']=='Mocha']['Calories'].mean()

print('Daily calorie :', round((morning+noon+evening)/3,2))
```

```python
round(df.groupby('Beverage')['Calories'].mean(),2)
```

```python
beverage_data = {
                'Beverage': list(df.groupby('Beverage')['Calories'].mean().index),
                'Calories': list(df.groupby('Beverage')['Calories'].mean().values)
                 }

df = pd.DataFrame(beverage_data)

# Create a dictionary to map beverage numbers to their calorie values
beverage_calories = dict(zip(range(1, len(df) + 1), df['Calories']))

# Get user input for the number of beverages consumed
num_beverages = int(input("How many times do you consume beverages daily? "))

# Display the available beverages with numbers for user selection
print("Available beverages:")
for num, beverage in enumerate(df['Beverage'], start=1):
    print(f"{num}. {beverage}")

# Loop to get details of each beverage
daily_calories = 0
for i in range(num_beverages):
    # Get user input for the selected beverage number
    beverage_num = int(input(f"Enter the number of beverage {i + 1}: "))

    # Check if the entered beverage number exists in the dictionary
    if beverage_num in beverage_calories:
        daily_calories += beverage_calories[beverage_num]
    else:
        print(f"Beverage number {beverage_num} not found in the database. Please select a valid number.")

# Display the result
print(f"\nTotal daily calories: {daily_calories} calories")
```
