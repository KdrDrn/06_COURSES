## 1. conda environment'ları listelemek:

```python
import subprocess

# `conda info -e` komutunu çalıştırın
result = subprocess.run(['conda', 'info', '-e'], capture_output=True, text=True)

# Çıktıyı ayırın
environment_name = result.stdout.split('\n')[2:]

# Ortam adını yazdırın
display("conda environment'lar:", environment_name)
```

## 2. aktif olan conda environment'ı döndürmek:

```python
result = subprocess.run(['conda', 'info', '--envs'], capture_output=True, text=True)
env_info = result.stdout
# Aktif environment'i bulmak için satır satır kontrol edelim
for line in env_info.split('\n'):
    if '*' in line:  # Aktif environment '*' ile işaretlenmiştir.
                    # Bu satırı parçalayarak sadece environment adını alabiliriz.
        current_env = line.split()[0]
print("aktif conda environment :", current_env)
```

## 3. aktif olan conda environment'ının package listesini döndürmek:

```python
# "conda list" komutunun çıktısını içeren bir string döndürelim.
# yani aktif olan conda environment'ının package listesini döndüreceğiz: 

result = subprocess.run(['conda', 'list'], capture_output=True, text=True)

package_list = result.stdout
package_list
```

```python
# string halindeki bu package bilgilerini liste içine alalım:

print("conda env. package listesi:\n")
package_list.split('\n')
```

```python
# package'lar ve versiyonları listenin 2. indeksinden itibaren başlıyor:

print("conda env. package listesi:\n")
package_list.split('\n')[2:]
```

```python
# package listesine kolay erişim için onu bir dictionary içine alalım:

package_lines = package_list.strip().split('\n')[2:]

installed_packages = {}
        
# Her satırı işleyerek package adını ve versiyonunu alacağız:
for line in package_lines:
    parts = line.split()
    package_name = parts[0]
    package_version = parts[1]
    installed_packages[package_name] = package_version

print("conda package listesi(dictionary içinde) :\n\n", installed_packages)
```

```python
# bu dictionary'nin "python" key'ini çağırarak python versiyonunu görelim:

print("\n\nconda env. python versiyonu :", installed_packages["python"])
```

## 6. tüm bu kodları user defined fonksiyonlarda birleştirelim:

```python
import subprocess
import sys
import subprocess

# Anaconda'nın veya miniconda'nın kurulu olduğu yerde bulunan
# conda komutunu kullanarak environment bilgisini alabilirsiniz.
# Bu kodlar Anaconda veya Miniconda'da çalışır.

def get_current_environment():
    try:
        result = subprocess.run(['conda', 'info', '--envs'], capture_output=True, text=True)
        env_info = result.stdout
        # Aktif environment'i bulmak için satır satır kontrol edelim
        for line in env_info.split('\n'):
            if '*' in line:
                # Aktif environment '*' ile işaretlenmiştir.
                # Bu satırı parçalayarak sadece environment adını alabiliriz.
                current_env = line.split()[0]
                return current_env
    except FileNotFoundError:
        print("Anaconda veya Miniconda yüklü değil veya 'conda' komutu bulunamadı.")
        return None




def get_installed_packages():
    try:
        # 'conda list' komutunu kullanarak kurulu paketlerin listesini alalım
        result = subprocess.run(['conda', 'list'], capture_output=True, text=True)
        package_list = result.stdout
        
        # İlk satır genellikle sütun başlıklarını içerir, onu atlayalım
        package_lines = package_list.strip().split('\n')[2:]
        
        # Paketlerin adlarını ve versiyonlarını depolamak için bir sözlük oluşturalım
        installed_packages = {}
        
        # Her satırı işleyerek paket adını ve versiyonunu alalım
        for line in package_lines:
            parts = line.split()
            package_name = parts[0]
            package_version = parts[1]
            installed_packages[package_name] = package_version
        
        return installed_packages
    except FileNotFoundError:
        print("Anaconda veya Miniconda yüklü değil veya 'conda' komutu bulunamadı.")
        return None
```

```python
# Aktif environment'i alalım:

if current_environment:
    display("Şu anda çalıştığınız environment: ", get_current_environment())
else:
    display("Environment bilgisi alınamadı.")

    
# Kurulu paketleri alalım:

installed_packages = get_installed_packages()

if installed_packages:
    print("Kurulu Paketler:")
    for package, version in installed_packages.items():
        print(f"{package}: {version}")
else:
    print("Kurulu paketler alınamadı.")

    
# Jupyter Notebook'un çalıştığı environment'ı görüntüleyelim:

print("\n\nJupyter Notebook'un çalıştığı environment:", sys.executable)


# jupyter notebook'un kullandığı python versiyonu:    

import sys
print("jupyter notebook'un şu an üzerinde çalıştığı environment'in python sürümü :",sys.version[:7],
      "\naktif olan conda environment'in python sürümü :", get_installed_packages()["python"])
```
