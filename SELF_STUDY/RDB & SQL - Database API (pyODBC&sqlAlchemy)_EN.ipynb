**What is an API?**

An API, or Application Programming Interface, is a set of definitions and protocols that allow two pieces of software to communicate with each other. APIs can be used for web applications, operating systems, databases, hardware, or software libraries.

**How do APIs work?**

APIs work by providing a way for two software components to exchange data and instructions. APIs act as an interface that allows one piece of software to access and use the functionality of another piece of software.

**Why are APIs important?**

APIs are an essential tool for allowing software to communicate with each other and exchange data. APIs make the software development process easier and allow software to be more efficient and flexible.

**APIs and Jupyter Notebook**

Jupyter Notebook is a popular tool for developing data analysis and machine learning applications with Python. Jupyter Notebook allows you to connect to SQL Server with the Python pyodbc module and write queries.

**Connecting to SQL Server with APIs**

To connect to SQL Server in Jupyter Notebook, you can use the `pyodbc` module. The pyodbc module provides a set of functions and classes that allow you to connect to SQL Server from Python.

**The connect()** function is a function used to create a connection to a SQL server and a database. This function takes the following parameters:

The only parameter of this function is a `connection string`. A connection string is a text string that contains the address of the SQL server, the username, password, and other connection information.

**A `connection string` can contain the following components:**

**Driver:** The name of the driver used to connect to the SQL server (and therefore the database).

**Server:** Specifies the name or IP address of the SQL server to connect to. This indicates where the database is hosted.

**Database:** Specifies the database you want to connect to. This indicates which database you want to access on the database server.

**UID:** Username.

**PWD:** Password.

**Port:** The port number of the SQL server.

**Timeout:** The timeout period for the connection attempt.

**Attributes:** Other properties to be applied to the connection.

**The part "TRUSTED_CONNECTION=yes;"** specifies that a trusted connection will be used. This means that the connection will be made securely using Windows credentials.

The expression ***"TRUSTED_CONNECTION=no;** specifies that a trusted connection will not be used. In this case, credentials such as username and password must be specified when making the connection."**

For example, the following connection string is used to connect to the "test" database running on "localhost" with the username "user" and password "password":

**`connection_string = "DRIVER={SQL Server};SERVER=localhost;DATABASE=test;UID=user;PWD=password"`**

If the connect() function is successful, it returns a pyodbc.Connection object. This object can be used to connect to the SQL server and receive and send data.**

**`The conn.autocommit = True`** command enables the automatic commit feature of a database connection. When the automatic commit feature is enabled, every change is automatically committed to the database.

If the automatic commit feature is disabled, the **`conn.commit()`** command must be used to commit changes to the database.

Database transactions typically consist of a series of SQL statements that make changes to the database by executing them together. However, these changes do not usually happen immediately; instead, they are accumulated in a "transaction". The **`conn.commit()`**  call ensures that the changes made within a database transaction are permanently committed. In other words, if all SQL statements within the transaction were successfully executed and no error occurred, these changes become permanent with the commit call. If an error occurred and the commit call was not made, the transaction can be rolled back and the changes in the database are not applied.

When using **`conn.commit()`**, you must eventually close the connection using the **`conn.close()`** command.

## The BEGIN TRANSACTION command in SQL Server:

```python
# start:
'BEGIN TRANSACTION'

# The queries should be here.
UPDATE products
SET price = price * 1.10
WHERE category = 'Electronics'

# End the transaction:
COMMIT TRANSACTION
```

In this example, the BEGIN TRANSACTION command starts a transaction. Then, the UPDATE command is used to increase the prices of products in the electronics category by 10%. Finally, the COMMIT TRANSACTION command is used to commit the transaction.

The BEGIN TRANSACTION command is important for maintaining data integrity. All changes made by a transaction are either committed successfully or rolled back completely. This ensures that transactions are consistent and reliable.**

**If we want to run the above SQL query in Python:**

(NOTE: The code is an example code. It will not work as it is since there is no such database.)

## If we use autocommit = False

```python
import pyodbc

conn_string = "driver={ODBC Driver 17 for SQL Server}; server=DESKTOP-O8V793J\SQLEXPRESS; database=master; TRUSTED_CONNECTION=yes;"

conn = pyodbc.connect(conn_string)
crs = conn.cursor()

conn.autocommit = False

# execute CREATE DATABASE command:
crs.execute("CREATE DATABASE TEST_C")  # ERROR !!!

conn.commit()  # This command is necessary for recording to the database because conn.autocommit is set to False

# close the connection:
conn.close()  # This command is necessary because conn.autocommit is set to False
```

As you can see, we get an error when we run a critical query like CREATE DATABASE with autocommit = False.

It fails because the `.execute()` method starts a transaction and `CREATE DATABASE` cannot be run within a transaction.

When you open a connection with `autocommit = False`, every operation is executed within a transaction. The CREATE DATABASE statement, on the other hand, cannot be executed within a transaction. Therefore, you will get an error when you try to run the `CREATE DATABASE` statement when you have opened a connection with `autocommit = False`.

SQL Server does not allow the `CREATE DATABASE` statement to be executed within such a transaction; therefore, the connection must be in autocommit mode before issuing such statements.

When a connection is opened with `autocommit = True`, the `CREATE DATABASE` statement does not start a transaction because it is automatically committed. Therefore, the `CREATE DATABASE` statement does not fail when a connection is opened with autocommit = True.

In summary, when a connection is opened with `autocommit = True`, every operation is automatically committed. This eliminates the issue of the `CREATE DATABASE` statement starting a transaction and not being able to be run within a transaction.

### <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">`pd.to_sql()` by using `sqlalchemy`</p>

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" 
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

**to_sql:** The to_sql function, a method of the Pandas DataFrame object, is used to transfer the contents of a DataFrame to a SQL table.

**name='your_table':** This parameter specifies the name of the SQL table to be created or overwritten. You can replace the expression 'your_table' with your own table name.

**con=conn:** This parameter specifies the SQLAlchemy engine and connection. conn is a connection object that connects to a SQL database through an SQLAlchemy engine.

**if_exists='replace':** This parameter specifies what to do if a table with the specified name already exists. 'replace' means to overwrite the existing table. Other options include 'fail' (throw an error), and 'append' (add the data to the existing table).

**index=False:** This parameter specifies not to include the index column of the DataFrame in the SQL table. If index=True is set, the index column of the DataFrame is added to the SQL table as a column.

**This code creates or overwrites a SQL table with the specified name and transfers the contents of the DataFrame to that table.**
